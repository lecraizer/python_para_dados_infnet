## Exercício 1 — Controle de Temperatura

### Conceitos trabalhados
- `input`
- `float`
- `if / elif / else`
- comparação com intervalos

### Enunciado
Um sistema simples monitora a temperatura de um ambiente.

Peça ao usuário uma temperatura e informe:
- `Frio`, se a temperatura for menor que 18
- `Agradável`, se estiver entre 18 e 26
- `Quente`, se for maior que 26

### Observação didática
Este exercício é simples, mas é útil para revisar **intervalos de valores**, algo em que muitos alunos ainda se confundem.

```python
temperatura = float(input("Digite a temperatura: "))

if temperatura < 18:
    print("Frio")
elif temperatura <= 26:
    print("Agradável")
else:
    print("Quente")
```

## Exercício 2 — Radar de Velocidade

### Conceitos trabalhados
- condicionais
- operadores relacionais
- classificação por faixas

### Enunciado
Um radar detecta a velocidade de um carro.

Classifique a situação da seguinte forma:

## Exercício 3 — Filtro de Lista

### Conceitos trabalhados
- listas
- `for`
- `append()`
- condição dentro de laço

### Enunciado
Peça 6 números ao usuário e crie uma nova lista contendo **apenas os números pares**.

No final, mostre a lista original e a lista filtrada.

### Observação didática
Aqui aparece uma ideia importante: **filtrar dados**. Esse tipo de raciocínio reaparece bastante depois em análise de dados.

```python
numeros = []
pares = []

for i in range(6):
    n = int(input(f"Digite o número {i+1}: "))
    numeros.append(n)

for numero in numeros:
    if numero % 2 == 0:
        pares.append(numero)

print("Lista original:", numeros)
print("Lista de pares:", pares)
```

## Exercício 4 — Senha com Validação

### Conceitos trabalhados
- `while`
- `len()`
- condição de parada

### Enunciado
Peça uma senha ao usuário até que ela tenha pelo menos **6 caracteres**.

Quando a senha atender ao critério, o programa deve encerrar com a mensagem `Senha válida!`.

### Observação didática
Esse exercício já se aproxima de um comportamento de sistema real.

```python
while True:
    senha = input("Digite uma senha: ")

    if len(senha) >= 6:
        print("Senha válida!")
        break
    else:
        print("Senha muito curta. Tente novamente.")
```

## Exercício 5 — Caixa do Mercado

### Conceitos trabalhados
- `while True`
- acumulador
- contador
- condição de parada

### Enunciado
Simule um caixa de mercado.

O usuário deve digitar os preços dos produtos comprados. O programa deve continuar lendo valores até que o usuário digite `0`.

Ao final, mostre:
- o valor total da compra
- a quantidade de produtos informados

### Observação didática
Esse é um ótimo exercício para reforçar dois padrões muito importantes:
- **contador**
- **acumulador**

```python
total = 0
quantidade = 0

while True:
    preco = float(input("Digite o preço do produto (0 para encerrar): "))

    if preco == 0:
        break

    total += preco
    quantidade += 1

print("Total da compra:", total)
print("Quantidade de produtos:", quantidade)
```

## Exercício 6 — Maior Valor Digitado

### Conceitos trabalhados
- `while`
- comparação
- atualização de maior valor

### Enunciado
Peça números ao usuário até que ele digite `-1`.

Ao final, mostre o maior número digitado.

### Observação didática
Este é um padrão clássico e muito importante: acompanhar o **maior valor** ao longo da execução.

```python
maior = None

while True:
    numero = int(input("Digite um número (-1 para encerrar): "))

    if numero == -1:
        break

    if maior is None or numero > maior:
        maior = numero

print("Maior número digitado:", maior)
```

## Exercício 7 — Contador de Letras Específicas

### Conceitos trabalhados
- strings
- `for` em string
- contagem com condição
- `lower()`

### Enunciado
Peça uma frase ao usuário e conte quantas vezes aparece a letra `a`.

### Observação didática
Apesar de simples, esse exercício conecta bem com a ideia de percorrer textos caractere por caractere.

```python
frase = input("Digite uma frase: ").lower()

contador = 0

for letra in frase:
    if letra == "a":
        contador += 1

print("Quantidade de letras 'a':", contador)
```

## Exercício 8 — Jogo de Adivinhação

### Conceitos trabalhados
- `while`
- condicionais
- módulo `random`
- repetição até acertar

### Enunciado
O computador escolhe um número aleatório entre 1 e 10.

O usuário deve tentar adivinhar. A cada tentativa, o programa informa se o número correto é maior ou menor. O jogo termina quando o usuário acerta.

### Observação didática
Esse exercício costuma engajar bem e mistura vários conceitos básicos em um problema curto.

```python
import random

numero_secreto = random.randint(1, 10)

while True:
    chute = int(input("Digite seu chute: "))

    if chute == numero_secreto:
        print("Acertou!")
        break
    elif chute < numero_secreto:
        print("O número correto é maior.")
    else:
        print("O número correto é menor.")
```

## Exercício 9 — Sistema de Notas

### Conceitos trabalhados
- `while`
- acumulador
- contador
- condição dentro do laço
- média

### Enunciado
Um professor quer analisar notas lançadas manualmente.

O programa deve:
- ler notas até que o usuário digite `-1`
- ao final, mostrar:
  - a média das notas
  - quantas notas foram maiores ou iguais a 7

### Observação didática
Esse exercício integra bem ideias importantes de entrada repetida, processamento e saída.

```python
soma = 0
quantidade = 0
aprovados = 0

while True:
    nota = float(input("Digite uma nota (-1 para encerrar): "))

    if nota == -1:
        break

    soma += nota
    quantidade += 1

    if nota >= 7:
        aprovados += 1

if quantidade > 0:
    media = soma / quantidade
else:
    media = 0

print("Média das notas:", media)
print("Quantidade de notas maiores ou iguais a 7:", aprovados)
```

## Exercício 10 — Mini Torneio entre 4 Jogadores

### Conceitos trabalhados
- listas
- `for` aninhado
- comparação
- sorteio com `random`
- funções
- divisão do problema em etapas

### Enunciado
Você vai desenvolver um programa que simula um mini torneio entre 4 jogadores.

O programa deve:

1. Ler o nome de 4 jogadores
2. Fazer com que cada jogador jogue contra todos os outros
3. Em cada partida:
   - cada jogador recebe uma pontuação aleatória de 0 a 10
   - quem fizer mais pontos vence e ganha 3 pontos
   - em caso de empate, cada jogador ganha 1 ponto
4. Ao final, o programa deve mostrar:
   - a pontuação final de cada jogador
   - o nome do campeão do torneio

### Requisito de organização
Organize o programa com:
- uma função principal
- funções auxiliares para partes importantes do processo, como:
  - ler os jogadores
  - simular uma partida
  - mostrar a pontuação final
  - descobrir o campeão

### Observação didática
Este exercício é excelente para revisar `for` aninhado e mostrar a utilidade real de dividir o código em funções.

```python
import random


def ler_jogadores():
    jogadores = []

    for i in range(4):
        nome = input(f"Digite o nome do jogador {i+1}: ")
        jogadores.append(nome)

    return jogadores


def simular_partida(jogador1, jogador2):
    pontos1 = random.randint(0, 10)
    pontos2 = random.randint(0, 10)

    print(f"{jogador1} ({pontos1}) x ({pontos2}) {jogador2}")

    if pontos1 > pontos2:
        return 3, 0
    elif pontos2 > pontos1:
        return 0, 3
    else:
        return 1, 1


def jogar_torneio(jogadores):
    pontuacoes = [0] * len(jogadores)

    for i in range(len(jogadores)):
        for j in range(i + 1, len(jogadores)):
            p1, p2 = simular_partida(jogadores[i], jogadores[j])
            pontuacoes[i] += p1
            pontuacoes[j] += p2

    return pontuacoes


def mostrar_resultado(jogadores, pontuacoes):
    print("\nPontuação final:")
    for i in range(len(jogadores)):
        print(f"{jogadores[i]}: {pontuacoes[i]} pontos")


def encontrar_campeao(jogadores, pontuacoes):
    maior_pontuacao = max(pontuacoes)
    indice = pontuacoes.index(maior_pontuacao)
    return jogadores[indice]


def main():
    jogadores = ler_jogadores()
    pontuacoes = jogar_torneio(jogadores)
    mostrar_resultado(jogadores, pontuacoes)

    campeao = encontrar_campeao(jogadores, pontuacoes)
    print(f"\nCampeão do torneio: {campeao}")


main()
```

## Exercício 11 — Ranking de Notas da Turma

### Conceitos trabalhados
- listas
- `for`
- média
- busca de maior valor
- contagem com condição
- funções auxiliares

### Enunciado
Um professor quer analisar as notas de uma turma.

Faça um programa que:

1. Leia o nome e a nota de 5 alunos
2. Armazene essas informações
3. Ao final, mostre:
   - a média da turma
   - o nome do aluno com a maior nota
   - quantos alunos ficaram com nota maior ou igual a 7
   - quantos alunos ficaram acima da média da turma

### Requisito de organização
Organize o programa com:
- uma função principal
- funções auxiliares para tarefas como:
  - ler os dados dos alunos
  - calcular a média
  - encontrar o aluno com maior nota
  - contar aprovados
  - contar quantos ficaram acima da média

### Observação didática
Esse exercício tem cara de problema real e é ótimo para consolidar a ideia de quebrar um problema maior em subproblemas menores.

```python
def ler_alunos():
    nomes = []
    notas = []

    for i in range(5):
        nome = input(f"Digite o nome do aluno {i+1}: ")
        nota = float(input(f"Digite a nota de {nome}: "))

        nomes.append(nome)
        notas.append(nota)

    return nomes, notas


def calcular_media(notas):
    return sum(notas) / len(notas)


def encontrar_melhor_aluno(nomes, notas):
    maior_nota = max(notas)
    indice = notas.index(maior_nota)
    return nomes[indice], maior_nota


def contar_aprovados(notas):
    contador = 0

    for nota in notas:
        if nota >= 7:
            contador += 1

    return contador


def contar_acima_da_media(notas, media):
    contador = 0

    for nota in notas:
        if nota > media:
            contador += 1

    return contador


def mostrar_resultados(nomes, notas):
    media = calcular_media(notas)
    melhor_aluno, melhor_nota = encontrar_melhor_aluno(nomes, notas)
    aprovados = contar_aprovados(notas)
    acima_da_media = contar_acima_da_media(notas, media)

    print("\n--- Resultado da turma ---")
    print(f"Média da turma: {media:.2f}")
    print(f"Melhor aluno: {melhor_aluno} ({melhor_nota})")
    print(f"Alunos com nota maior ou igual a 7: {aprovados}")
    print(f"Alunos acima da média: {acima_da_media}")


def main():
    nomes, notas = ler_alunos()
    mostrar_resultados(nomes, notas)


main()
```
