## Exercício 1 - Mesclar dois dicionários somando valores

Some os valores de chaves que aparecem nos dois dicionários. Chaves exclusivas de um dos lados devem aparecer normalmente no resultado.

```python
janeiro = {"ana": 300, "bruno": 150, "carol": 200}
fevereiro = {"bruno": 100, "carol": 250, "daniel": 400}

resultado = dict(janeiro)
for nome, valor in fevereiro.items():
    resultado[nome] = resultado.get(nome, 0) + valor

print(resultado)
```

## Exercício 2 - Filtrar produtos em promoção

Use list comprehension para retornar apenas os produtos com desconto acima de 20%.

```python
produtos = [
    {"nome": "camisa", "preco": 100, "desconto": 0.10},
    {"nome": "tênis", "preco": 200, "desconto": 0.30},
    {"nome": "calça", "preco": 150, "desconto": 0.25},
    {"nome": "meia", "preco": 30, "desconto": 0.05}
]

promocao = [p for p in produtos if p["desconto"] > 0.20]
print(promocao)
```

## Exercício 3 - Extrair campo específico

Use list comprehension para extrair apenas os nomes dos clientes que fizeram mais de 2 pedidos.

```python
clientes = [
    {"nome": "Ana", "pedidos": 5},
    {"nome": "Bruno", "pedidos": 1},
    {"nome": "Carol", "pedidos": 3},
    {"nome": "Daniel", "pedidos": 2}
]

ativos = [c["nome"] for c in clientes if c["pedidos"] > 2]
print(ativos)
```

## Exercício 4 - Atualizar saldo com histórico

Dado um saldo inicial e uma lista de movimentos (positivos e negativos), calcule o saldo final e guarde o histórico de saldos após cada movimento.

```python
saldo = 1000
movimentos = [200, -150, -300, 500, -100]

historico = []
for m in movimentos:
    saldo += m
    historico.append(saldo)

saldo_final = saldo

print("Histórico:", historico)
print("Saldo final:", saldo_final)
```

## Exercício 5 - Agrupar transações por status

Agrupe as transações pelo campo `status`, somando o valor total de cada grupo.

```python
transacoes = [
    {"id": 1, "status": "aprovado", "valor": 300},
    {"id": 2, "status": "pendente", "valor": 150},
    {"id": 3, "status": "aprovado", "valor": 200},
    {"id": 4, "status": "recusado", "valor": 400},
    {"id": 5, "status": "pendente", "valor": 100}
]

totais = {}
for t in transacoes:
    status = t["status"]
    totais[status] = totais.get(status, 0) + t["valor"]

print(totais)
```
