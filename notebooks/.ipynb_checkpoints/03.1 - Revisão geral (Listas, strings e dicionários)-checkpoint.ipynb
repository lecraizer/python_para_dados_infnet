# Revisão de Python: Listas, Strings e Dicionários

Este notebook é uma revisão prática de três estruturas/conceitos fundamentais em Python:

- **listas**, com foco em criação, acesso, alteração e métodos como `append`, `remove`, `extend`, `insert`, `pop`, `sort`, entre outros;
- **strings**, com foco em indexação, fatiamento e métodos importantes como `lower`, `upper`, `strip`, `replace`, `split`, `join`, `startswith`, `endswith`, `find` e outros;
- **dicionários**, com foco em criação, acesso, atualização, remoção, iteração e aplicações comuns como contagem de frequência e agrupamento de dados.

A ideia é revisar a lógica de uso dessas ferramentas e praticar com exemplos próximos de situações reais.

## 1. Listas

Uma **lista** é uma estrutura que armazena vários valores em uma única variável.

As listas em Python:

- são ordenadas;
- aceitam elementos repetidos;
- podem guardar diferentes tipos de dados;
- são mutáveis, isto é, podem ser alteradas depois de criadas.

````python
nomes = ["Ana", "Bruno", "Carla", "Diego"]
idades = [20, 25, 19, 31]
misturada = ["Ana", 20, True, 8.5]

nomes
````

### 1.1 Acessando elementos

A contagem dos índices começa em **0**.

````python
nomes = ["Ana", "Bruno", "Carla", "Diego"]

print(nomes[0])   # primeiro elemento
print(nomes[1])   # segundo elemento
print(nomes[-1])  # último elemento
````

### 1.2 Fatiamento de listas

O fatiamento permite pegar uma parte da lista.

A lógica geral é:

```python
lista[inicio:fim]
```

O índice `fim` não entra no resultado.

````python
numeros = [10, 20, 30, 40, 50, 60]

print(numeros[0:3])
print(numeros[2:5])
print(numeros[:4])
print(numeros[3:])
print(numeros[::-1])  # lista invertida
````

### 1.3 Alterando valores de uma lista

Como listas são mutáveis, podemos modificar elementos diretamente pelo índice.

````python
notas = [7.5, 8.0, 6.5]
notas[2] = 9.0

notas
````

## 2. Métodos importantes de listas

Métodos são ações que uma lista sabe executar.

A forma geral é:

```python
lista.metodo()
```

### 2.1 `append`: adicionar um elemento ao final

````python
frutas = ["banana", "maçã"]
frutas.append("uva")

frutas
````

### 2.2 `extend`: adicionar vários elementos ao final

Enquanto `append` adiciona um único elemento, `extend` adiciona os elementos de outra sequência.

````python
frutas = ["banana", "maçã"]
frutas.extend(["uva", "laranja", "abacaxi"])

frutas
````

Compare com o uso de `append` com uma lista:

````python
frutas = ["banana", "maçã"]
frutas.append(["uva", "laranja"])

frutas
````

### 2.3 `insert`: inserir em uma posição específica

````python
nomes = ["Ana", "Carla", "Diego"]
nomes.insert(1, "Bruno")

nomes
````

### 2.4 `remove`: remover pelo valor

O método `remove` remove a primeira ocorrência do valor informado.

````python
produtos = ["camisa", "calça", "camisa", "tênis"]
produtos.remove("camisa")

produtos
````

### 2.5 `pop`: remover pelo índice

Se nenhum índice for informado, `pop` remove o último elemento.

````python
produtos = ["camisa", "calça", "tênis", "boné"]

removido = produtos.pop(1)

print("Produto removido:", removido)
print("Lista final:", produtos)
````

````python
produtos = ["camisa", "calça", "tênis", "boné"]
ultimo = produtos.pop()

print("Último removido:", ultimo)
print(produtos)
````

### 2.6 `clear`: limpar a lista

````python
itens = ["a", "b", "c"]
itens.clear()

itens
````

### 2.7 `index`: descobrir a posição de um elemento

````python
nomes = ["Ana", "Bruno", "Carla", "Diego"]
posicao = nomes.index("Carla")

posicao
````

### 2.8 `count`: contar ocorrências

````python
respostas = ["sim", "não", "sim", "sim", "não"]
qtd_sim = respostas.count("sim")

qtd_sim
````

### 2.9 `sort` e `sorted`: ordenar listas

`sort` altera a própria lista.

`sorted` retorna uma nova lista ordenada.

````python
notas = [8.5, 6.0, 9.5, 7.0]
notas.sort()

notas
````

````python
notas = [8.5, 6.0, 9.5, 7.0]
ordenadas = sorted(notas)

print("Original:", notas)
print("Ordenada:", ordenadas)
````

````python
notas = [8.5, 6.0, 9.5, 7.0]
notas.sort(reverse=True)

notas
````

### 2.10 `reverse`: inverter a ordem da lista

````python
valores = [1, 2, 3, 4, 5]
valores.reverse()

valores
````

### 2.11 Resumo dos principais métodos de listas

| Método | O que faz |
|---|---|
| `append(x)` | adiciona `x` ao final da lista |
| `extend(lista)` | adiciona vários elementos ao final |
| `insert(i, x)` | insere `x` na posição `i` |
| `remove(x)` | remove a primeira ocorrência de `x` |
| `pop(i)` | remove e retorna o elemento da posição `i` |
| `clear()` | remove todos os elementos |
| `index(x)` | retorna a posição de `x` |
| `count(x)` | conta quantas vezes `x` aparece |
| `sort()` | ordena a própria lista |
| `reverse()` | inverte a ordem da própria lista |

## 3. Percorrendo listas com `for`

Uma das operações mais comuns é percorrer todos os elementos de uma lista.

````python
nomes = ["Ana", "Bruno", "Carla"]

for nome in nomes:
    print("Olá,", nome)
````

### 3.1 Usando `enumerate`

O `enumerate` permite acessar, ao mesmo tempo, o índice e o valor.

````python
nomes = ["Ana", "Bruno", "Carla"]

for indice, nome in enumerate(nomes):
    print(indice, nome)
````

### 3.2 Acumulando valores

````python
notas = [7.5, 8.0, 6.5, 9.0]
soma = 0

for nota in notas:
    soma += nota

media = soma / len(notas)
media
````

## 4. Strings

Uma **string** é uma sequência de caracteres.

Em Python, strings podem ser acessadas por índice e fatiadas de forma parecida com listas.

````python
texto = "Python"

print(texto[0])
print(texto[-1])
print(texto[0:3])
````

### 4.1 Strings são imutáveis

Diferente das listas, strings não podem ser alteradas diretamente por índice.

````python
palavra = "gato"

# Isto daria erro:
# palavra[0] = "r"

nova_palavra = "r" + palavra[1:]
nova_palavra
````

## 5. Métodos importantes de strings

### 5.1 `lower`, `upper`, `title` e `capitalize`

````python
nome = "ana maria"

print(nome.lower())
print(nome.upper())
print(nome.title())
print(nome.capitalize())
````

### 5.2 `strip`, `lstrip` e `rstrip`

Esses métodos removem espaços em branco das pontas da string.

````python
texto = "   Python é legal   "

print(texto.strip())
print(texto.lstrip())
print(texto.rstrip())
````

### 5.3 `replace`: substituir partes do texto

````python
frase = "Eu gosto de Java"
frase = frase.replace("Java", "Python")

frase
````

### 5.4 `split`: quebrar uma string em uma lista

````python
frase = "banana,maçã,uva,laranja"
frutas = frase.split(",")

frutas
````

### 5.5 `join`: juntar elementos de uma lista em uma string

````python
frutas = ["banana", "maçã", "uva"]
texto = " - ".join(frutas)

texto
````

### 5.6 `startswith` e `endswith`

Esses métodos verificam se uma string começa ou termina com determinado trecho.

````python
arquivo = "relatorio_final.pdf"

print(arquivo.startswith("relatorio"))
print(arquivo.endswith(".pdf"))
````

### 5.7 `find` e `index`

Ambos procuram um trecho dentro da string.

A diferença principal é:

- `find` retorna `-1` se não encontrar;
- `index` gera erro se não encontrar.

````python
email = "aluno@faculdade.com"

print(email.find("@"))
print(email.find("gmail"))
````

### 5.8 `count`: contar ocorrências em uma string

````python
frase = "banana"
frase.count("a")
````

### 5.9 `in`: verificar se um texto aparece em outro

````python
email = "aluno@faculdade.com"

if "@" in email:
    print("E-mail contém arroba")
else:
    print("E-mail inválido")
````

### 5.10 Métodos úteis para validação

Alguns métodos retornam `True` ou `False` de acordo com o conteúdo da string.

````python
print("123".isdigit())
print("abc".isalpha())
print("abc123".isalnum())
print("   ".isspace())
````

### 5.11 Resumo dos principais métodos de strings

| Método | O que faz |
|---|---|
| `lower()` | transforma em minúsculas |
| `upper()` | transforma em maiúsculas |
| `title()` | coloca iniciais em maiúsculas |
| `capitalize()` | coloca a primeira letra em maiúscula |
| `strip()` | remove espaços das pontas |
| `replace(a, b)` | troca `a` por `b` |
| `split(sep)` | quebra a string em uma lista |
| `join(lista)` | junta elementos em uma string |
| `startswith(x)` | verifica se começa com `x` |
| `endswith(x)` | verifica se termina com `x` |
| `find(x)` | procura `x` e retorna a posição ou `-1` |
| `count(x)` | conta ocorrências de `x` |
| `isdigit()` | verifica se contém apenas dígitos |
| `isalpha()` | verifica se contém apenas letras |
| `isalnum()` | verifica se contém letras e/ou números |

## 6. Dicionários

Um **dicionário** armazena dados no formato **chave: valor**.

Ele é muito útil quando queremos associar uma informação a outra.

````python
aluno = {
    "nome": "Ana",
    "idade": 22,
    "curso": "Python",
    "ativo": True
}

aluno
````

### 6.1 Acessando valores

````python
aluno = {
    "nome": "Ana",
    "idade": 22,
    "curso": "Python"
}

print(aluno["nome"])
print(aluno["curso"])
````

### 6.2 Usando `get`

O método `get` é útil porque não gera erro se a chave não existir.

````python
aluno = {
    "nome": "Ana",
    "idade": 22
}

print(aluno.get("nome"))
print(aluno.get("email"))
print(aluno.get("email", "E-mail não cadastrado"))
````

### 6.3 Alterando e adicionando valores

````python
aluno = {
    "nome": "Ana",
    "idade": 22
}

aluno["idade"] = 23
aluno["curso"] = "Python"

aluno
````

### 6.4 Removendo elementos

````python
aluno = {
    "nome": "Ana",
    "idade": 22,
    "curso": "Python"
}

curso_removido = aluno.pop("curso")

print("Curso removido:", curso_removido)
print(aluno)
````

````python
aluno = {
    "nome": "Ana",
    "idade": 22,
    "curso": "Python"
}

del aluno["idade"]

aluno
````

### 6.5 Verificando se uma chave existe

````python
aluno = {
    "nome": "Ana",
    "idade": 22
}

if "idade" in aluno:
    print("A chave idade existe")
````

## 7. Percorrendo dicionários

### 7.1 Percorrendo as chaves

````python
aluno = {
    "nome": "Ana",
    "idade": 22,
    "curso": "Python"
}

for chave in aluno:
    print(chave)
````

### 7.2 Percorrendo os valores

````python
for valor in aluno.values():
    print(valor)
````

### 7.3 Percorrendo chave e valor ao mesmo tempo

````python
for chave, valor in aluno.items():
    print(chave, "->", valor)
````

### 7.4 Resumo dos principais métodos de dicionários

| Método | O que faz |
|---|---|
| `get(chave)` | acessa uma chave sem gerar erro se ela não existir |
| `keys()` | retorna as chaves |
| `values()` | retorna os valores |
| `items()` | retorna pares chave-valor |
| `pop(chave)` | remove uma chave e retorna seu valor |
| `update(outro_dict)` | atualiza/adiciona várias chaves |
| `clear()` | limpa o dicionário |

## 8. Aplicações comuns com dicionários

### 8.1 Contagem de frequência

Uma aplicação muito comum de dicionários é contar quantas vezes cada item aparece.

````python
frutas = ["banana", "maçã", "banana", "uva", "maçã", "banana"]
contagem = {}

for fruta in frutas:
    if fruta not in contagem:
        contagem[fruta] = 1
    else:
        contagem[fruta] += 1

contagem
````

### 8.2 Agrupando dados

Também podemos usar dicionários para agrupar registros por categoria.

````python
vendas = [
    {"produto": "camisa", "categoria": "roupa", "valor": 80},
    {"produto": "calça", "categoria": "roupa", "valor": 150},
    {"produto": "mouse", "categoria": "eletrônico", "valor": 90},
    {"produto": "teclado", "categoria": "eletrônico", "valor": 120},
]

soma_por_categoria = {}

for venda in vendas:
    categoria = venda["categoria"]
    valor = venda["valor"]
    
    if categoria not in soma_por_categoria:
        soma_por_categoria[categoria] = 0
    
    soma_por_categoria[categoria] += valor

soma_por_categoria
````

### 8.3 Lista de dicionários

Em muitos problemas reais, os dados aparecem como uma lista de dicionários.

````python
alunos = [
    {"nome": "Ana", "nota": 8.5, "turma": "A"},
    {"nome": "Bruno", "nota": 6.0, "turma": "B"},
    {"nome": "Carla", "nota": 9.0, "turma": "A"},
    {"nome": "Diego", "nota": 7.0, "turma": "B"},
]

for aluno in alunos:
    if aluno["nota"] >= 7:
        print(aluno["nome"], "foi aprovado(a)")
````

## Exercício 2 — Controle de notas

Dada a lista abaixo:

```python
notas = [7.5, 8.0, 6.0, 9.5, 5.5]
```

Calcule a média da turma e crie uma nova lista chamada `aprovadas` contendo apenas as notas maiores ou iguais a 7.

### Gabarito

````python
notas = [7.5, 8.0, 6.0, 9.5, 5.5]

media = sum(notas) / len(notas)
aprovadas = []

for nota in notas:
    if nota >= 7:
        aprovadas.append(nota)

print("Média:", media)
print("Notas aprovadas:", aprovadas)
````

## Exercício 6 — Validação simples de e-mails


## Exercício 9 — Soma por categoria

```python
vendas = [
```

```python
produtos = [
```

Encontre o nome do produto mais caro.

# Desafio final

Você recebeu uma lista de cadastros de alunos. Cada aluno é representado por um dicionário.

Tarefas:

1. tratar os nomes usando `strip` e `title`;
2. criar uma nova chave `status`, indicando `"aprovado"` para nota maior ou igual a 7 e `"reprovado"` caso contrário;
3. contar quantos alunos foram aprovados e reprovados;
4. criar uma lista apenas com os e-mails válidos, considerando válidos os que contêm `@` e terminam com `.com`.

````python
alunos = [
    {"nome": " ana ", "nota": 8.5, "email": "ana@gmail.com"},
    {"nome": "BRUNO", "nota": 6.0, "email": "bruno.com"},
    {"nome": " carla", "nota": 9.0, "email": "carla@faculdade.com"},
    {"nome": "DIEGO ", "nota": 5.5, "email": "diego@"},
    {"nome": "eduardo", "nota": 7.0, "email": "eduardo@hotmail.com"},
]
````

### Gabarito do desafio final

````python
aprovados = 0
reprovados = 0
emails_validos = []

for aluno in alunos:
    aluno["nome"] = aluno["nome"].strip().title()
    
    if aluno["nota"] >= 7:
        aluno["status"] = "aprovado"
        aprovados += 1
    else:
        aluno["status"] = "reprovado"
        reprovados += 1
    
    if "@" in aluno["email"] and aluno["email"].endswith(".com"):
        emails_validos.append(aluno["email"])

print("Alunos tratados:")
print(alunos)
print()
print("Aprovados:", aprovados)
print("Reprovados:", reprovados)
print("E-mails válidos:", emails_validos)
````
