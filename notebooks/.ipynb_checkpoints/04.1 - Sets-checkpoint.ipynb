# Revisão de Python: Conjuntos (`set`)

Este notebook é uma revisão focada em **conjuntos** em Python, isto é, o tipo `set`.

A ideia é entender:

- o que é um conjunto em Python;
- quando faz sentido usar `set` em vez de lista;
- como criar conjuntos;
- como adicionar, remover e verificar elementos;
- quais são as principais operações: união, interseção, diferença e diferença simétrica;
- como `set` se compara com `list`, `tuple` e `dict`;
- alguns cuidados importantes de sintaxe e uso.

Conjuntos são muito úteis quando queremos trabalhar com **elementos únicos**, ou seja, sem repetição.

## 1. O que é um `set`?

Um `set` é uma coleção de elementos com duas características centrais:

1. **Não permite elementos repetidos**.
2. **Não possui ordem fixa garantida para acesso por índice**.

Ou seja, um conjunto serve muito bem quando queremos responder perguntas como:

- Quais valores diferentes existem em uma lista?
- Um elemento pertence a um grupo?
- Quais elementos aparecem em duas coleções ao mesmo tempo?
- Quais elementos aparecem em uma coleção, mas não em outra?

Em Python, conjuntos são escritos usando `{}` com elementos separados por vírgula.

````python
frutas = {'maçã', 'banana', 'uva', 'banana'}
frutas
````

Observe que `'banana'` foi escrita duas vezes, mas aparece apenas uma vez no conjunto.

Isso acontece porque o `set` elimina duplicatas automaticamente.

## 2. Comparando `list`, `set` e `dict`

Visualmente, `set` e `dict` podem confundir um pouco porque ambos usam chaves `{}`.

A diferença está no conteúdo:

```python
# Lista
[1, 2, 3]

# Set / conjunto
{1, 2, 3}

# Dicionário
{'nome': 'Ana', 'idade': 20}
```

A lista usa colchetes `[]`.

O conjunto usa chaves `{}`, mas contém apenas valores.

O dicionário também usa chaves `{}`, mas contém pares `chave: valor`.

````python
lista = ['Ana', 'Bruno', 'Carlos']
conjunto = {'Ana', 'Bruno', 'Carlos'}
dicionario = {'nome': 'Ana', 'idade': 20}

print(type(lista))
print(type(conjunto))
print(type(dicionario))
````

## 3. Cuidado: `{}` vazio não cria um set

Um detalhe importante: em Python, `{}` cria um **dicionário vazio**, não um conjunto vazio.

Para criar um conjunto vazio, usamos `set()`.

````python
vazio_1 = {}
vazio_2 = set()

print(type(vazio_1))
print(type(vazio_2))
````

## 4. Criando conjuntos a partir de listas

Uma das aplicações mais comuns de `set` é remover valores repetidos de uma lista.

````python
nomes = ['Ana', 'Bruno', 'Ana', 'Carlos', 'Bruno', 'Daniela']
nomes_unicos = set(nomes)

nomes_unicos
````

Se quisermos voltar para lista, podemos usar `list()`.

Mas atenção: a ordem original pode não ser preservada, porque conjuntos não são estruturas ordenadas para acesso por índice.

````python
lista_sem_repeticao = list(set(nomes))
lista_sem_repeticao
````

## 5. Set não usa índice

Em listas, podemos acessar elementos por posição:

```python
lista[0]
```

Em conjuntos, isso não funciona, porque um `set` não é pensado como uma sequência ordenada.

````python
cores = {'azul', 'verde', 'vermelho'}

# A linha abaixo daria erro:
# cores[0]
````

## 6. Verificando se um elemento está no conjunto

Uma das grandes vantagens de `set` é verificar pertencimento com `in`.

Isso é muito comum em filtros, validações e remoção de duplicatas.

````python
usuarios_bloqueados = {'ana', 'carlos', 'joao'}

usuario = 'ana'

if usuario in usuarios_bloqueados:
    print('Usuário bloqueado')
else:
    print('Usuário liberado')
````

Também podemos usar `not in`.

````python
usuario = 'mariana'

if usuario not in usuarios_bloqueados:
    print('Usuário não está bloqueado')
````

## 7. Adicionando elementos: `add()`

Para adicionar um único elemento ao conjunto, usamos `.add()`.

````python
linguagens = {'Python', 'JavaScript'}
linguagens.add('SQL')

linguagens
````

Se adicionarmos um elemento que já existe, o conjunto continua igual.

````python
linguagens.add('Python')
linguagens
````

## 8. Adicionando vários elementos: `update()`

Para adicionar vários elementos de uma vez, usamos `.update()`.

Ele recebe uma coleção, como lista, tupla ou outro conjunto.

````python
linguagens = {'Python', 'SQL'}
linguagens.update(['R', 'Julia', 'Python'])

linguagens
````

Note que `'Python'` não foi duplicado.

## 9. Diferença entre `add()` e `update()`

- `add()` adiciona **um elemento**.
- `update()` adiciona **vários elementos**, percorrendo uma coleção.

Cuidado com strings: uma string também é uma coleção de caracteres.

````python
letras = set()
letras.add('abc')
print(letras)

letras = set()
letras.update('abc')
print(letras)
````

No primeiro caso, `'abc'` entra como uma única string.

No segundo caso, cada caractere da string é adicionado separadamente.

## 10. Removendo elementos: `remove()`

O método `.remove()` remove um elemento do conjunto.

Mas existe um detalhe importante: se o elemento não existir, ocorre erro.

````python
numeros = {1, 2, 3, 4}
numeros.remove(3)

numeros
````

````python
# A linha abaixo daria erro, pois 10 não está no conjunto:
# numeros.remove(10)
````

## 11. Removendo sem erro: `discard()`

O método `.discard()` também remove um elemento, mas não dá erro se ele não existir.

````python
numeros = {1, 2, 3, 4}
numeros.discard(10)

numeros
````

Comparação prática:

- `remove(x)`: remove `x`, mas dá erro se `x` não existir.
- `discard(x)`: remove `x`, se existir; se não existir, não faz nada.

## 12. Removendo um elemento qualquer: `pop()`

O método `.pop()` remove e retorna um elemento do conjunto.

Como o conjunto não tem uma ordem de índice, não devemos pensar em `.pop()` como “remover o último”, como acontece em listas.

````python
valores = {'a', 'b', 'c'}
removido = valores.pop()

print('Removido:', removido)
print('Conjunto final:', valores)
````

## 13. Limpando o conjunto: `clear()`

O método `.clear()` remove todos os elementos do conjunto.

````python
valores = {10, 20, 30}
valores.clear()

valores
````

## 14. Tamanho de um conjunto: `len()`

Assim como em listas, strings e dicionários, usamos `len()` para contar quantos elementos existem em um conjunto.

````python
produtos = {'camisa', 'calça', 'sapato', 'camisa'}
len(produtos)
````

Apesar de termos escrito `'camisa'` duas vezes, ela conta apenas uma vez.

## 15. Percorrendo um conjunto com `for`

Podemos percorrer um conjunto com `for`, mas não devemos depender da ordem de exibição.

````python
cores = {'azul', 'verde', 'vermelho'}

for cor in cores:
    print(cor)
````

Se a ordem for importante, talvez uma lista seja uma estrutura mais adequada.

# 16. Operações de conjuntos

A parte mais importante de `set` aparece quando comparamos dois grupos de elementos.

As principais operações são:

- união;
- interseção;
- diferença;
- diferença simétrica.

Vamos imaginar dois grupos de alunos matriculados em cursos diferentes.

````python
python = {'Ana', 'Bruno', 'Carlos', 'Daniela'}
sql = {'Carlos', 'Daniela', 'Eduardo', 'Fernanda'}

print('Python:', python)
print('SQL:', sql)
````

## 17. União

A união junta os elementos dos dois conjuntos, sem repetir.

Em termos de pergunta:

> Quem está em Python **ou** em SQL?

````python
alunos_total = python.union(sql)
alunos_total
````

Também podemos usar o operador `|`.

````python
python | sql
````

## 18. Interseção

A interseção retorna os elementos que aparecem nos dois conjuntos.

Em termos de pergunta:

> Quem está em Python **e também** em SQL?

````python
alunos_nos_dois = python.intersection(sql)
alunos_nos_dois
````

Também podemos usar o operador `&`.

````python
python & sql
````

## 19. Diferença

A diferença retorna os elementos que estão no primeiro conjunto, mas não no segundo.

Em termos de pergunta:

> Quem está em Python, mas não está em SQL?

````python
so_python = python.difference(sql)
so_python
````

Também podemos usar o operador `-`.

````python
python - sql
````

A ordem importa na diferença.

````python
print('Python - SQL:', python - sql)
print('SQL - Python:', sql - python)
````

## 20. Diferença simétrica

A diferença simétrica retorna os elementos que estão em apenas um dos conjuntos, mas não nos dois ao mesmo tempo.

Em termos de pergunta:

> Quem está em apenas um dos cursos?

````python
apenas_um_curso = python.symmetric_difference(sql)
apenas_um_curso
````

Também podemos usar o operador `^`.

````python
python ^ sql
````

## 21. Resumo visual das operações

Usando os conjuntos:

```python
python = {'Ana', 'Bruno', 'Carlos', 'Daniela'}
sql = {'Carlos', 'Daniela', 'Eduardo', 'Fernanda'}
```

Temos:

| Operação | Pergunta | Resultado esperado |
|---|---|---|
| `python | sql` | Quem está em Python ou SQL? | Todos, sem repetição |
| `python & sql` | Quem está nos dois? | Carlos e Daniela |
| `python - sql` | Quem está só em Python? | Ana e Bruno |
| `sql - python` | Quem está só em SQL? | Eduardo e Fernanda |
| `python ^ sql` | Quem está em apenas um dos dois? | Ana, Bruno, Eduardo e Fernanda |

# 22. Relações entre conjuntos

Além de gerar novos conjuntos, também podemos fazer perguntas sobre a relação entre conjuntos.

As principais são:

- um conjunto é subconjunto de outro?
- um conjunto contém outro?
- os conjuntos não têm nenhum elemento em comum?

## 23. Subconjunto: `issubset()`

Um conjunto é subconjunto de outro quando todos os seus elementos estão no outro conjunto.

````python
alunos_python = {'Ana', 'Bruno', 'Carlos'}
aprovados = {'Ana', 'Bruno'}

aprovados.issubset(alunos_python)
````

Também podemos usar o operador `<=`.

````python
aprovados <= alunos_python
````

## 24. Superconjunto: `issuperset()`

Um conjunto é superconjunto de outro quando ele contém todos os elementos do outro.

````python
alunos_python.issuperset(aprovados)
````

Também podemos usar o operador `>=`.

````python
alunos_python >= aprovados
````

## 25. Conjuntos disjuntos: `isdisjoint()`

Dois conjuntos são disjuntos quando não possuem nenhum elemento em comum.

````python
grupo_a = {'Ana', 'Bruno'}
grupo_b = {'Carlos', 'Daniela'}
grupo_c = {'Bruno', 'Eduardo'}

print(grupo_a.isdisjoint(grupo_b))
print(grupo_a.isdisjoint(grupo_c))
````

# 26. Métodos que alteram o próprio conjunto

Alguns métodos criam um novo conjunto como resultado.

Outros alteram o próprio conjunto original.

Por exemplo:

- `union()` cria um novo conjunto;
- `update()` altera o conjunto original;
- `intersection_update()` mantém no conjunto original apenas a interseção;
- `difference_update()` remove do conjunto original os elementos de outro;
- `symmetric_difference_update()` atualiza o conjunto com a diferença simétrica.

## 27. `update()` como união acumulada

````python
a = {1, 2, 3}
b = {3, 4, 5}

a.update(b)
a
````

## 28. `intersection_update()`

````python
a = {1, 2, 3}
b = {2, 3, 4}

a.intersection_update(b)
a
````

## 29. `difference_update()`

````python
a = {1, 2, 3, 4}
b = {2, 4}

a.difference_update(b)
a
````

## 30. `symmetric_difference_update()`

````python
a = {1, 2, 3}
b = {3, 4, 5}

a.symmetric_difference_update(b)
a
````

# 31. Conversões úteis

Podemos converter entre listas, tuplas e conjuntos.

Isso é útil principalmente quando queremos remover duplicatas ou fazer operações de comparação.

````python
lista = [1, 2, 2, 3, 3, 3, 4]
conjunto = set(lista)
lista_nova = list(conjunto)

tprint = print
print('Lista original:', lista)
print('Conjunto:', conjunto)
print('Lista nova:', lista_nova)
````

## 32. Removendo duplicatas preservando ordem

Um detalhe importante: `list(set(lista))` remove duplicatas, mas pode alterar a ordem.

Se quisermos remover duplicatas preservando a primeira aparição, podemos usar uma lógica com `for`.

````python
nomes = ['Ana', 'Bruno', 'Ana', 'Carlos', 'Bruno', 'Daniela']

vistos = set()
resultado = []

for nome in nomes:
    if nome not in vistos:
        resultado.append(nome)
        vistos.add(nome)

resultado
````

Aqui usamos o `set` como estrutura auxiliar para verificar rapidamente se um nome já apareceu.

# 33. Set com strings

Quando aplicamos `set()` diretamente em uma string, o Python cria um conjunto com os caracteres únicos da string.

````python
palavra = 'banana'
set(palavra)
````

Isso pode ser útil para verificar letras diferentes em uma palavra.

````python
palavra = 'programacao'
letras_unicas = set(palavra)

print(letras_unicas)
print(len(letras_unicas))
````

# 34. Set e dicionários

Como `set` e `dict` usam `{}`, vale reforçar a diferença.

Um conjunto contém elementos:

```python
{'Ana', 'Bruno', 'Carlos'}
```

Um dicionário contém pares `chave: valor`:

```python
{'Ana': 8.5, 'Bruno': 7.0, 'Carlos': 9.2}
```

Além disso, as chaves de um dicionário funcionam de forma parecida com um conjunto: elas também não podem se repetir.

````python
notas = {'Ana': 8.5, 'Bruno': 7.0, 'Ana': 9.0}
notas
````

No exemplo acima, a chave `'Ana'` apareceu duas vezes. O valor final associado a ela foi o último informado.

Isso lembra a ideia do `set`: chaves de dicionário também são únicas.

## 35. Transformando chaves de dicionário em set

Às vezes queremos comparar chaves de dicionários.

````python
notas_turma_a = {'Ana': 8.5, 'Bruno': 7.0, 'Carlos': 9.0}
notas_turma_b = {'Carlos': 8.0, 'Daniela': 9.5, 'Eduardo': 6.5}

alunos_a = set(notas_turma_a.keys())
alunos_b = set(notas_turma_b.keys())

print('Alunos nas duas turmas:', alunos_a & alunos_b)
print('Alunos apenas na turma A:', alunos_a - alunos_b)
print('Alunos apenas na turma B:', alunos_b - alunos_a)
````

# 36. Set comprehension

Assim como listas podem ser criadas com list comprehension, conjuntos também podem ser criados com set comprehension.

A sintaxe é parecida, mas usamos `{}`.

````python
numeros = [1, 2, 2, 3, 4, 4, 5]
quadrados = {n ** 2 for n in numeros}

quadrados
````

Como o resultado é um conjunto, valores repetidos são eliminados.

````python
palavras = ['casa', 'carro', 'casa', 'cidade', 'carro']
tamanhos = {len(palavra) for palavra in palavras}

tamanhos
````

# 37. O que pode entrar em um set?

Elementos de um conjunto precisam ser **hashable**, isto é, precisam ser tipos que o Python consiga usar como elementos únicos de forma estável.

Na prática, normalmente podemos colocar:

- números;
- strings;
- tuplas imutáveis;
- booleanos.

Mas não podemos colocar listas ou dicionários diretamente dentro de um set, porque são estruturas mutáveis.

````python
valores = {1, 2.5, 'Python', True, (10, 20)}
valores
````

````python
# A linha abaixo daria erro:
# conjunto = {[1, 2], [3, 4]}
````

## 38. Por que listas não podem entrar em sets?

Listas podem mudar depois de criadas.

Como o `set` precisa controlar elementos únicos, ele precisa de elementos estáveis.

Por isso, uma tupla pode entrar em um set, mas uma lista não.

````python
pontos = {(1, 2), (3, 4), (1, 2)}
pontos
````

# 39. `frozenset`

Existe também o `frozenset`, que é uma versão imutável de `set`.

Ele funciona como um conjunto, mas não permite adicionar ou remover elementos depois de criado.

É menos comum em aulas iniciais, mas vale conhecer.

````python
conjunto_congelado = frozenset([1, 2, 3, 3])
conjunto_congelado
````

````python
# A linha abaixo daria erro, pois frozenset é imutável:
# conjunto_congelado.add(4)
````

# 40. Quando usar `set`?

Use `set` quando você precisa de:

- remover duplicatas;
- verificar rapidamente se um elemento pertence a uma coleção;
- comparar grupos de elementos;
- calcular elementos em comum;
- calcular elementos exclusivos;
- representar categorias únicas.

Evite usar `set` quando:

- a ordem dos elementos importa;
- você precisa acessar elementos por índice;
- elementos repetidos têm significado;
- você precisa associar chaves a valores, caso em que um dicionário é mais adequado.

# 41. Exemplos pontuais

A seguir estão alguns exemplos curtos de uso de conjuntos.

## Exemplo 1: remover duplicatas de uma lista

````python
produtos = ['camisa', 'calça', 'camisa', 'tênis', 'calça']
produtos_unicos = set(produtos)

produtos_unicos
````

## Exemplo 2: verificar categorias válidas

````python
categorias_validas = {'roupa', 'calçado', 'acessório'}
categoria = 'roupa'

if categoria in categorias_validas:
    print('Categoria válida')
else:
    print('Categoria inválida')
````

## Exemplo 3: encontrar alunos em comum

````python
turma_manha = {'Ana', 'Bruno', 'Carlos'}
turma_tarde = {'Carlos', 'Daniela', 'Eduardo'}

comuns = turma_manha & turma_tarde
comuns
````

## Exemplo 4: encontrar permissões faltantes

````python
permissoes_necessarias = {'ler', 'escrever', 'editar'}
permissoes_usuario = {'ler', 'editar'}

faltantes = permissoes_necessarias - permissoes_usuario
faltantes
````

## Exemplo 5: letras únicas em uma frase

````python
frase = 'python para dados'
letras = set(frase.replace(' ', ''))

letras
````

# 42. Mini-revisão final

Principais pontos:

- `set` representa conjuntos de elementos únicos.
- Não permite duplicatas.
- Não é acessado por índice.
- Usa `{}` com valores separados por vírgula.
- `{}` vazio cria dicionário, não conjunto.
- Para conjunto vazio, usamos `set()`.
- `add()` adiciona um elemento.
- `update()` adiciona vários elementos.
- `remove()` remove, mas dá erro se o elemento não existir.
- `discard()` remove sem erro se o elemento não existir.
- `union()` / `|` faz união.
- `intersection()` / `&` faz interseção.
- `difference()` / `-` faz diferença.
- `symmetric_difference()` / `^` faz diferença simétrica.
- `issubset()`, `issuperset()` e `isdisjoint()` verificam relações entre conjuntos.
