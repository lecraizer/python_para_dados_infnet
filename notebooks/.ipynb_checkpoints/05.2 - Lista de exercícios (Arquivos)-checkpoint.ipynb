# Lista de Exercícios — Abertura de Arquivos em Python

Esta lista contém 10 exercícios sobre abertura, leitura e escrita de arquivos em Python.

A proposta é praticar:

- criar arquivos com Python;
- abrir arquivos com `open()`;
- usar `with open(...)`;
- ler arquivos com `.read()` e com `for linha in arquivo`;
- escrever arquivos com `.write()`;
- transformar conteúdo de arquivos em listas e dicionários;
- fazer parsing simples de arquivos `.txt` e `.csv`;
- aplicar listas, strings, dicionários e sets junto com arquivos.

Cada exercício possui gabarito logo abaixo.


## Exercício 1 — Criando um arquivo de anotações

Crie um arquivo chamado `anotacoes.txt` contendo três linhas:

```text
Estudar arquivos em Python
Praticar leitura de arquivos
Resolver exercícios
```

Depois, abra o arquivo e mostre seu conteúdo na tela.


````python
# Gabarito

with open("anotacoes.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write("Estudar arquivos em Python\n")
    arquivo.write("Praticar leitura de arquivos\n")
    arquivo.write("Resolver exercícios\n")

with open("anotacoes.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

print(conteudo)
````

## Exercício 2 — Transformando linhas de um arquivo em lista

Crie um arquivo chamado `compras.txt` com os seguintes itens, um por linha:

```text
arroz
feijão
macarrão
café
leite
```

Depois, leia esse arquivo e crie uma lista chamada `itens`, contendo os nomes dos produtos sem `\n`.


````python
# Gabarito

with open("compras.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write("arroz\n")
    arquivo.write("feijão\n")
    arquivo.write("macarrão\n")
    arquivo.write("café\n")
    arquivo.write("leite\n")

itens = []

with open("compras.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        itens.append(linha.strip())

itens
````

## Exercício 3 — Acrescentando conteúdo ao final do arquivo

Use o arquivo `compras.txt` criado no exercício anterior.

Acrescente os itens abaixo ao final do arquivo:

```text
açúcar
óleo
```

Depois, leia novamente o arquivo e mostre todos os itens.


````python
# Gabarito

with open("compras.txt", "a", encoding="utf-8") as arquivo:
    arquivo.write("açúcar\n")
    arquivo.write("óleo\n")

with open("compras.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        print(linha.strip())
````

## Exercício 4 — Contando linhas de um arquivo

Crie uma função chamada `contar_linhas(nome_arquivo)`.

A função deve abrir um arquivo de texto e retornar quantas linhas ele possui.

Teste a função usando o arquivo `compras.txt`.


````python
# Gabarito

def contar_linhas(nome_arquivo):
    quantidade = 0

    with open(nome_arquivo, "r", encoding="utf-8") as arquivo:
        for linha in arquivo:
            quantidade += 1

    return quantidade

contar_linhas("compras.txt")
````

## Exercício 5 — Buscando palavras dentro de um arquivo

Crie um arquivo chamado `frases.txt` com o seguinte conteúdo:

```text
Python é uma linguagem muito usada.
Arquivos podem guardar dados.
Ler arquivos com Python é útil.
Dicionários ajudam a organizar informações.
```

Depois, crie uma função chamada `buscar_linhas(nome_arquivo, palavra)`.

A função deve retornar uma lista com as linhas que contêm a palavra buscada.

Teste buscando a palavra `"Python"`.


````python
# Gabarito

with open("frases.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write("Python é uma linguagem muito usada.\n")
    arquivo.write("Arquivos podem guardar dados.\n")
    arquivo.write("Ler arquivos com Python é útil.\n")
    arquivo.write("Dicionários ajudam a organizar informações.\n")

def buscar_linhas(nome_arquivo, palavra):
    encontradas = []

    with open(nome_arquivo, "r", encoding="utf-8") as arquivo:
        for linha in arquivo:
            if palavra.lower() in linha.lower():
                encontradas.append(linha.strip())

    return encontradas

buscar_linhas("frases.txt", "Python")
````

## Exercício 6 — Criando um relatório simples a partir de um arquivo

Use o arquivo `frases.txt`.

Crie uma função chamada `relatorio_texto(nome_arquivo)` que retorna um dicionário com:

- `"linhas"`: quantidade de linhas;
- `"palavras"`: quantidade total de palavras;
- `"caracteres"`: quantidade total de caracteres, desconsiderando quebras de linha.


````python
# Gabarito

def relatorio_texto(nome_arquivo):
    total_linhas = 0
    total_palavras = 0
    total_caracteres = 0

    with open(nome_arquivo, "r", encoding="utf-8") as arquivo:
        for linha in arquivo:
            linha_limpa = linha.strip()

            total_linhas += 1
            total_palavras += len(linha_limpa.split())
            total_caracteres += len(linha_limpa)

    return {
        "linhas": total_linhas,
        "palavras": total_palavras,
        "caracteres": total_caracteres
    }

relatorio_texto("frases.txt")
````

## Exercício 7 — Criando e lendo um CSV de tipos sanguíneos

Crie um arquivo chamado `tipos_sanguineos.csv` com as colunas:

```text
nome,tipo_sanguineo
```

Use os dados abaixo:

```text
Ana,A
Bruno,O
Carlos,B
Diana,A
Eduardo,AB
Fernanda,O
Gabriel,A
Helena,B
Igor,O
Juliana,AB
```

Depois, crie uma função chamada `contar_tipos_sanguineos(nome_arquivo)`.

A função deve abrir o CSV e retornar um dicionário com a contagem de cada tipo sanguíneo.


````python
# Gabarito

with open("tipos_sanguineos.csv", "w", encoding="utf-8") as arquivo:
    arquivo.write("nome,tipo_sanguineo\n")
    arquivo.write("Ana,A\n")
    arquivo.write("Bruno,O\n")
    arquivo.write("Carlos,B\n")
    arquivo.write("Diana,A\n")
    arquivo.write("Eduardo,AB\n")
    arquivo.write("Fernanda,O\n")
    arquivo.write("Gabriel,A\n")
    arquivo.write("Helena,B\n")
    arquivo.write("Igor,O\n")
    arquivo.write("Juliana,AB\n")

def contar_tipos_sanguineos(nome_arquivo):
    contagem = {}

    with open(nome_arquivo, "r", encoding="utf-8") as arquivo:
        next(arquivo)  # pula o cabeçalho

        for linha in arquivo:
            nome, tipo = linha.strip().split(",")
            contagem[tipo] = contagem.get(tipo, 0) + 1

    return contagem

contar_tipos_sanguineos("tipos_sanguineos.csv")
````

## Exercício 8 — Criando uma lista de dicionários a partir de CSV

Use o arquivo `tipos_sanguineos.csv`.

Crie uma função chamada `ler_pessoas(nome_arquivo)`.

A função deve retornar uma lista de dicionários no formato:

```python
[
    {"nome": "Ana", "tipo_sanguineo": "A"},
    {"nome": "Bruno", "tipo_sanguineo": "O"}
]
```


````python
# Gabarito

def ler_pessoas(nome_arquivo):
    pessoas = []

    with open(nome_arquivo, "r", encoding="utf-8") as arquivo:
        next(arquivo)

        for linha in arquivo:
            nome, tipo = linha.strip().split(",")

            pessoa = {
                "nome": nome,
                "tipo_sanguineo": tipo
            }

            pessoas.append(pessoa)

    return pessoas

ler_pessoas("tipos_sanguineos.csv")
````

## Exercício 9 — Filtrando registros e salvando em outro arquivo

Use o arquivo `tipos_sanguineos.csv`.

Crie um novo arquivo chamado `tipo_a.txt`, contendo apenas os nomes das pessoas com tipo sanguíneo `"A"`.

O arquivo final deve ficar assim:

```text
Ana
Diana
Gabriel
```


````python
# Gabarito

with open("tipos_sanguineos.csv", "r", encoding="utf-8") as entrada:
    next(entrada)

    with open("tipo_a.txt", "w", encoding="utf-8") as saida:
        for linha in entrada:
            nome, tipo = linha.strip().split(",")

            if tipo == "A":
                saida.write(nome + "\n")

with open("tipo_a.txt", "r", encoding="utf-8") as arquivo:
    print(arquivo.read())
````

## Exercício 10 — Limpando dados de um arquivo

Crie um arquivo chamado `emails.txt` com o seguinte conteúdo:

```text
 ANA@EMAIL.COM
bruno@email.com
ana@email.com
 carlos@email.com 
BRUNO@email.com
```

Depois, crie uma função chamada `limpar_emails(nome_arquivo)`.

A função deve:

- abrir o arquivo;
- remover espaços extras;
- transformar tudo em minúsculas;
- remover e-mails repetidos;
- retornar uma lista final de e-mails únicos.


````python
# Gabarito

with open("emails.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write(" ANA@EMAIL.COM\n")
    arquivo.write("bruno@email.com\n")
    arquivo.write("ana@email.com\n")
    arquivo.write(" carlos@email.com \n")
    arquivo.write("BRUNO@email.com\n")

def limpar_emails(nome_arquivo):
    emails = set()

    with open(nome_arquivo, "r", encoding="utf-8") as arquivo:
        for linha in arquivo:
            email = linha.strip().lower()
            emails.add(email)

    return list(emails)

limpar_emails("emails.txt")
````
