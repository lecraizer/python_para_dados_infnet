# Aula — JSON em Python

Este notebook é focado exclusivamente em **JSON**.

A ideia é entender:

- o que é JSON;
- por que JSON é usado;
- como JSON se parece com dicionários e listas em Python;
- como transformar JSON em estruturas Python;
- como transformar estruturas Python em JSON;
- como ler e salvar arquivos `.json`;
- como acessar e manipular dados vindos de JSON.

O foco aqui não é ainda uma aula geral de arquivos, mas sim o formato **JSON** e o módulo `json` do Python.


## 1. O que é JSON?

JSON significa **JavaScript Object Notation**.

Apesar do nome vir do JavaScript, JSON é usado em muitas linguagens de programação, incluindo Python.

Ele é um formato de texto usado para armazenar e transmitir dados estruturados.

JSON aparece muito em:

- APIs;
- arquivos de configuração;
- bancos de dados NoSQL;
- sistemas web;
- troca de dados entre programas;
- respostas de serviços na internet.

Um exemplo de JSON:

```json
{
    "nome": "Ana",
    "idade": 22,
    "curso": "Python"
}
```

Visualmente, ele lembra bastante um dicionário Python.


## 2. JSON parece dicionário, mas não é exatamente dicionário

Um JSON pode parecer assim:

```json
{
    "nome": "Ana",
    "idade": 22,
    "ativo": true
}
```

Em Python, algo parecido seria:

```python
{
    "nome": "Ana",
    "idade": 22,
    "ativo": True
}
```

A principal diferença é que JSON é um **formato de texto**.

Já o dicionário Python é uma **estrutura de dados da linguagem Python**.

Quando lemos um JSON no Python, normalmente convertemos esse texto para dicionários, listas, strings, números e booleanos.


## 3. Comparando JSON com Python

| JSON | Python |
|---|---|
| objeto `{}` | dicionário |
| lista `[]` | lista |
| string | string |
| número inteiro | int |
| número decimal | float |
| `true` | `True` |
| `false` | `False` |
| `null` | `None` |

Exemplo em JSON:

```json
{
    "nome": "Bruno",
    "idade": 25,
    "ativo": true,
    "telefone": null,
    "cursos": ["Python", "SQL"]
}
```

Depois de lido no Python, esse conteúdo vira uma combinação de dicionário, lista, string, inteiro, booleano e `None`.


## 4. O módulo `json`

Para trabalhar com JSON em Python, usamos o módulo `json`.

Ele já vem instalado com o Python.

Para importar:

```python
import json
```

As funções principais são:

| Função | Uso |
|---|---|
| `json.loads()` | transforma uma string JSON em objeto Python |
| `json.dumps()` | transforma objeto Python em string JSON |
| `json.load()` | lê JSON de um arquivo |
| `json.dump()` | salva objeto Python em arquivo JSON |


````python
import json
````

## 5. Lendo uma string JSON com `json.loads()`

Quando temos um texto em formato JSON, usamos `json.loads()`.

A letra `s` pode ser lembrada como **string**.

Observe que o conteúdo abaixo é uma string, não um dicionário Python.


````python
texto_json = '{"nome": "Ana", "idade": 22, "curso": "Python"}'

print(texto_json)
print(type(texto_json))
````

Agora vamos transformar essa string JSON em um dicionário Python.


````python
dados = json.loads(texto_json)

print(dados)
print(type(dados))
````

Depois da conversão, podemos acessar os dados como em um dicionário comum.


````python
print(dados["nome"])
print(dados["idade"])
print(dados["curso"])
````

## 6. JSON com listas

JSON também pode conter listas.

Exemplo:


````python
texto_json = '''
{
    "nome": "Carlos",
    "idade": 28,
    "cursos": ["Python", "SQL", "Power BI"]
}
'''

dados = json.loads(texto_json)

dados
````

A chave `"cursos"` virou uma lista em Python.


````python
print(dados["cursos"])
print(type(dados["cursos"]))
print(dados["cursos"][0])
````

## 7. JSON com lista de dicionários

Muitas vezes, um JSON contém uma lista de registros.

Esse formato é muito comum quando recebemos dados de uma API ou de um sistema.

Exemplo:


````python
texto_json = '''
[
    {"nome": "Ana", "idade": 22, "curso": "Python"},
    {"nome": "Bruno", "idade": 25, "curso": "SQL"},
    {"nome": "Carlos", "idade": 20, "curso": "Python"}
]
'''

alunos = json.loads(texto_json)

alunos
````

````python
print(type(alunos))
print(type(alunos[0]))
````

Como agora temos uma lista de dicionários, podemos usar `for`.


````python
for aluno in alunos:
    print(aluno["nome"], "-", aluno["curso"])
````

Também podemos filtrar os dados.


````python
alunos_python = []

for aluno in alunos:
    if aluno["curso"] == "Python":
        alunos_python.append(aluno)

alunos_python
````

## 8. Transformando Python em JSON com `json.dumps()`

A função `json.dumps()` transforma um objeto Python em uma string JSON.

Exemplo:


````python
produto = {
    "nome": "Camisa",
    "preco": 79.90,
    "em_estoque": True
}

texto_json = json.dumps(produto)

print(texto_json)
print(type(texto_json))
````

Observe que `True` virou `true`, seguindo o padrão JSON.


## 9. Melhorando a visualização com `indent`

O parâmetro `indent` deixa o JSON mais organizado visualmente.


````python
texto_json_formatado = json.dumps(produto, indent=4)

print(texto_json_formatado)
````

## 10. Preservando acentos com `ensure_ascii=False`

Por padrão, alguns caracteres especiais podem aparecer codificados.

Para preservar acentos, usamos `ensure_ascii=False`.


````python
aluno = {
    "nome": "João",
    "cidade": "Niterói",
    "curso": "Introdução ao Python"
}

texto_json = json.dumps(aluno, ensure_ascii=False, indent=4)

print(texto_json)
````

## 11. Lendo JSON de um arquivo com `json.load()`

Quando o JSON está salvo em um arquivo `.json`, usamos `json.load()`.

Para deixar o notebook completo, primeiro vamos criar um arquivo JSON de exemplo.


````python
aluno = {
    "nome": "Ana",
    "idade": 22,
    "curso": "Python para Dados",
    "ativo": True,
    "notas": [8.5, 9.0, 7.5]
}

with open("aluno.json", "w", encoding="utf-8") as arquivo:
    json.dump(aluno, arquivo, ensure_ascii=False, indent=4)
````

Agora vamos abrir esse arquivo e carregar os dados.


````python
with open("aluno.json", "r", encoding="utf-8") as arquivo:
    dados_aluno = json.load(arquivo)

dados_aluno
````

````python
print(type(dados_aluno))
print(dados_aluno["nome"])
print(dados_aluno["notas"])
````

## 12. Salvando JSON em arquivo com `json.dump()`

Quando queremos salvar um dicionário, lista ou outra estrutura Python em um arquivo JSON, usamos `json.dump()`.

A diferença principal:

- `json.dumps()` gera uma string;
- `json.dump()` grava em um arquivo.


````python
turma = [
    {"nome": "Ana", "curso": "Python", "nota": 8.5},
    {"nome": "Bruno", "curso": "SQL", "nota": 7.0},
    {"nome": "Carlos", "curso": "Python", "nota": 9.0},
    {"nome": "Diana", "curso": "Power BI", "nota": 8.0}
]

with open("turma.json", "w", encoding="utf-8") as arquivo:
    json.dump(turma, arquivo, ensure_ascii=False, indent=4)
````

Agora podemos ler esse arquivo de volta.


````python
with open("turma.json", "r", encoding="utf-8") as arquivo:
    turma_lida = json.load(arquivo)

turma_lida
````

## 13. Manipulando dados vindos de JSON

Depois que o JSON foi carregado, ele vira estrutura Python.

Por isso podemos calcular, filtrar, criar listas e dicionários normalmente.


### Exemplo: calcular média das notas

````python
soma = 0

for aluno in turma_lida:
    soma += aluno["nota"]

media = soma / len(turma_lida)

media
````

### Exemplo: criar lista com alunos de Python

````python
alunos_python = []

for aluno in turma_lida:
    if aluno["curso"] == "Python":
        alunos_python.append(aluno["nome"])

alunos_python
````

### Exemplo: contar alunos por curso

````python
contagem_cursos = {}

for aluno in turma_lida:
    curso = aluno["curso"]

    if curso not in contagem_cursos:
        contagem_cursos[curso] = 0

    contagem_cursos[curso] += 1

contagem_cursos
````

## 14. Diferença entre `load`, `loads`, `dump` e `dumps`

Essa é uma das partes mais importantes.

| Função | Entrada | Saída |
|---|---|---|
| `json.loads()` | string JSON | objeto Python |
| `json.dumps()` | objeto Python | string JSON |
| `json.load()` | arquivo JSON aberto | objeto Python |
| `json.dump()` | objeto Python + arquivo aberto | grava no arquivo |

Uma forma simples de lembrar:

- tem `s` no final: trabalha com **string**;
- não tem `s`: trabalha com **arquivo**.


## 15. Cuidados comuns com JSON

### 1. Aspas duplas

JSON usa aspas duplas para strings e chaves:

```json
{"nome": "Ana"}
```

Isto não é JSON válido:

```json
{'nome': 'Ana'}
```

Embora pareça um dicionário Python, não é JSON válido.

### 2. Booleanos

Em JSON:

```json
true
false
```

Em Python:

```python
True
False
```

### 3. Valor nulo

Em JSON:

```json
null
```

Em Python:

```python
None
```

### 4. Vírgula final

Em JSON, normalmente não se coloca vírgula depois do último item:

```json
{
    "nome": "Ana",
    "idade": 22
}
```


# Exemplos básicos finais

A seguir, alguns exemplos curtos para consolidar.


## Exemplo 1 — String JSON para dicionário

````python
texto = '{"produto": "Notebook", "preco": 3500}'

dados = json.loads(texto)

dados["produto"]
````

## Exemplo 2 — Dicionário para string JSON formatada

````python
cliente = {
    "nome": "Fernanda",
    "idade": 31,
    "cidade": "Rio de Janeiro"
}

print(json.dumps(cliente, ensure_ascii=False, indent=4))
````

## Exemplo 3 — Lista de dicionários para JSON

````python
produtos = [
    {"nome": "Camisa", "preco": 79.90},
    {"nome": "Calça", "preco": 149.90},
    {"nome": "Tênis", "preco": 299.90}
]

texto_json = json.dumps(produtos, ensure_ascii=False, indent=4)

print(texto_json)
````

## Exemplo 4 — Lendo arquivo JSON e acessando dados

````python
with open("turma.json", "r", encoding="utf-8") as arquivo:
    turma = json.load(arquivo)

for aluno in turma:
    print(aluno["nome"], aluno["nota"])
````

## Exemplo 5 — Criando um resumo a partir de JSON

````python
resumo = {
    "quantidade_alunos": len(turma),
    "media_notas": sum(aluno["nota"] for aluno in turma) / len(turma)
}

resumo
````
