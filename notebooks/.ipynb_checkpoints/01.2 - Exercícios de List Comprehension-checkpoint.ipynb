# 🧪 Exercícios — List Comprehension

## 1. Quadrados dos pares

Retorne os quadrados apenas dos números pares de uma lista.

```
def quadrados_pares(lista):
    return [n**2 for n in lista if n % 2 == 0]
```

## 2. Strings sem espaços

Remova os espaços extras de cada string.

```
def limpar_strings(lista):
    return [s.strip() for s in lista]
```

## 3. Palavras com mais de 3 letras

Retorne palavras com mais de 3 letras.

```
def palavras_longas(lista):
    return [p for p in lista if len(p) > 3]
```

## 4. Números positivos

Retorne apenas números positivos.

```
def positivos(lista):
    return [n for n in lista if n > 0]
```

## 5. Par ou ímpar

Retorne 'par' ou 'impar' para cada número.

```
def par_ou_impar(lista):
    return ["par" if n % 2 == 0 else "impar" for n in lista]
```

## 6. Substituir negativos

Substitua negativos por 0.

```
def substituir_negativos(lista):
    return [n if n >= 0 else 0 for n in lista]
```

## 7. Classificação de notas

Retorne 'aprovado' se >=7 senão 'reprovado'.

```
def classificar_notas(lista):
    return ["aprovado" if n >= 7 else "reprovado" for n in lista]
```

## 8. Primeira letra

Retorne a primeira letra de cada palavra.

```
def primeiras_letras(lista):
    return [p[0] for p in lista if p]
```

## 9. Palavras com 'a'

Retorne palavras que contêm a letra 'a'.

```
def contem_a(lista):
    return [p for p in lista if "a" in p.lower()]
```

## 10. Inverter palavras

Retorne cada palavra invertida.

```
def inverter_palavras(lista):
    return [p[::-1] for p in lista]
```

## 11. Remover vogais

Remova vogais de uma string.

```
def remover_vogais(texto):
    vogais = "aeiouAEIOU"
    return [c for c in texto if c not in vogais]
```

## 12. Pares como string

Retorne apenas pares convertidos para string.

```
def pares_str(lista):
    return [str(n) for n in lista if n % 2 == 0]
```

## 13. Começa com vogal

Retorne palavras que começam com vogal.

```
def comeca_vogal(lista):
    vogais = "aeiou"
    return [p for p in lista if p and p[0].lower() in vogais]
```

## 14. Substituir vazios

Substitua strings vazias por 'N/A'.

```
def tratar_vazios(lista):
    return [p if p != "" else "N/A" for p in lista]
```

## 15. Contar vogais

Conte o número de vogais em cada palavra.

```
def contar_vogais(lista):
    vogais = "aeiou"
    return [sum(1 for c in p.lower() if c in vogais) for p in lista]
```

## 16. Normalizar nomes

Primeira letra maiúscula e resto minúsculo.

```
def normalizar_nomes(lista):
    return [p.capitalize() for p in lista]
```

## 17. Acima da média

Retorne números maiores que a média.

```
def acima_media(lista):
    media = sum(lista) / len(lista)
    return [n for n in lista if n > media]
```
