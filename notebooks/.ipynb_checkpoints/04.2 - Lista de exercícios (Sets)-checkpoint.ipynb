# Lista de Exercícios — Sets em Python com Gabarito

Esta lista revisa o uso de **sets** (`set`) em Python de forma mais prática e um pouco mais desafiadora.

A ideia não é trabalhar conjuntos de forma muito matemática, mas usar `set` como ferramenta para resolver problemas comuns envolvendo:

- listas com valores repetidos;
- strings;
- dicionários;
- relatórios simples;
- comparação entre coleções;
- validação de dados;
- extração de informações únicas.

Cada exercício possui uma solução possível logo abaixo.


## Exercício 1 — Relatório de nomes únicos e repetidos

Você recebeu uma lista de nomes de alunos inscritos em uma oficina. Alguns nomes aparecem mais de uma vez.

```python
alunos = ["Ana", "Bruno", "Carlos", "Ana", "Diana", "Bruno", "Eduardo", "Ana"]
```

Crie um dicionário chamado `relatorio` contendo:

- `"total_inscricoes"`: quantidade total de itens na lista;
- `"total_alunos_unicos"`: quantidade de alunos diferentes;
- `"alunos_unicos"`: conjunto com os nomes únicos;
- `"houve_repeticao"`: `True` se houve algum nome repetido, `False` caso contrário.


````python
# Gabarito

alunos = ["Ana", "Bruno", "Carlos", "Ana", "Diana", "Bruno", "Eduardo", "Ana"]

alunos_unicos = set(alunos)

relatorio = {
    "total_inscricoes": len(alunos),
    "total_alunos_unicos": len(alunos_unicos),
    "alunos_unicos": alunos_unicos,
    "houve_repeticao": len(alunos) != len(alunos_unicos)
}

relatorio
````

## Exercício 2 — Identificando itens repetidos

Dada a lista abaixo, descubra quais códigos aparecem mais de uma vez.

```python
codigos = [101, 203, 304, 101, 502, 708, 203, 101]
```

Crie um conjunto chamado `codigos_repetidos`.

O resultado esperado é um conjunto contendo apenas os códigos que aparecem repetidos.


````python
# Gabarito

codigos = [101, 203, 304, 101, 502, 708, 203, 101]

vistos = set()
codigos_repetidos = set()

for codigo in codigos:
    if codigo in vistos:
        codigos_repetidos.add(codigo)
    else:
        vistos.add(codigo)

codigos_repetidos
````

## Exercício 3 — Palavras únicas em uma frase com tratamento simples

Dada a frase abaixo:

```python
frase = "Python é legal, python é útil e PYTHON aparece muito."
```

Crie um conjunto com as palavras únicas da frase, considerando:

- tudo em minúsculas;
- remoção de vírgula e ponto.

Depois, crie um dicionário com:

- `"palavras_unicas"`: conjunto das palavras únicas;
- `"quantidade"`: número de palavras diferentes.


````python
# Gabarito

frase = "Python é legal, python é útil e PYTHON aparece muito."

frase_limpa = frase.lower()
frase_limpa = frase_limpa.replace(",", "")
frase_limpa = frase_limpa.replace(".", "")

palavras = frase_limpa.split()
palavras_unicas = set(palavras)

resultado = {
    "palavras_unicas": palavras_unicas,
    "quantidade": len(palavras_unicas)
}

resultado
````

## Exercício 4 — Comparando presenças em duas aulas

Duas listas representam os alunos presentes em duas aulas.

```python
aula_1 = ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"]
aula_2 = ["Carlos", "Diana", "Fernanda", "Gabriel", "Ana"]
```

Crie um dicionário chamado `presencas` contendo:

- `"presentes_nas_duas"`: alunos presentes nas duas aulas;
- `"presentes_em_alguma"`: alunos presentes em pelo menos uma aula;
- `"apenas_aula_1"`: alunos que aparecem apenas na primeira aula;
- `"apenas_aula_2"`: alunos que aparecem apenas na segunda aula.


````python
# Gabarito

aula_1 = ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"]
aula_2 = ["Carlos", "Diana", "Fernanda", "Gabriel", "Ana"]

set_aula_1 = set(aula_1)
set_aula_2 = set(aula_2)

presencas = {
    "presentes_nas_duas": set_aula_1.intersection(set_aula_2),
    "presentes_em_alguma": set_aula_1.union(set_aula_2),
    "apenas_aula_1": set_aula_1.difference(set_aula_2),
    "apenas_aula_2": set_aula_2.difference(set_aula_1)
}

presencas
````

## Exercício 5 — Categorias únicas em produtos

Você recebeu uma lista de produtos, cada um representado por um dicionário.

```python
produtos = [
    {"nome": "Camisa Azul", "categoria": "camisa"},
    {"nome": "Camisa Preta", "categoria": "camisa"},
    {"nome": "Tênis Casual", "categoria": "calçado"},
    {"nome": "Meia Branca", "categoria": "meia"},
    {"nome": "Tênis Esportivo", "categoria": "calçado"},
    {"nome": "Boné", "categoria": "acessório"}
]
```

Crie um dicionário chamado `resumo` contendo:

- `"categorias"`: conjunto com as categorias únicas;
- `"quantidade_categorias"`: quantidade de categorias diferentes;
- `"quantidade_produtos"`: quantidade total de produtos.


````python
# Gabarito

produtos = [
    {"nome": "Camisa Azul", "categoria": "camisa"},
    {"nome": "Camisa Preta", "categoria": "camisa"},
    {"nome": "Tênis Casual", "categoria": "calçado"},
    {"nome": "Meia Branca", "categoria": "meia"},
    {"nome": "Tênis Esportivo", "categoria": "calçado"},
    {"nome": "Boné", "categoria": "acessório"}
]

categorias = set()

for produto in produtos:
    categorias.add(produto["categoria"])

resumo = {
    "categorias": categorias,
    "quantidade_categorias": len(categorias),
    "quantidade_produtos": len(produtos)
}

resumo
````

## Exercício 6 — Separando clientes novos e recorrentes

Uma loja registrou os clientes que compraram em janeiro e fevereiro.

```python
janeiro = ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"]
fevereiro = ["Carlos", "Eduardo", "Fernanda", "Gabriel", "Ana"]
```

Crie um dicionário chamado `clientes` contendo:

- `"recorrentes"`: clientes que compraram nos dois meses;
- `"novos_em_fevereiro"`: clientes que compraram em fevereiro, mas não em janeiro;
- `"nao_voltaram"`: clientes que compraram em janeiro, mas não em fevereiro.


````python
# Gabarito

janeiro = ["Ana", "Bruno", "Carlos", "Diana", "Eduardo"]
fevereiro = ["Carlos", "Eduardo", "Fernanda", "Gabriel", "Ana"]

set_janeiro = set(janeiro)
set_fevereiro = set(fevereiro)

clientes = {
    "recorrentes": set_janeiro.intersection(set_fevereiro),
    "novos_em_fevereiro": set_fevereiro.difference(set_janeiro),
    "nao_voltaram": set_janeiro.difference(set_fevereiro)
}

clientes
````

## Exercício 7 — Validando permissões de usuários

Um sistema exige as permissões abaixo para acessar uma área administrativa:

```python
permissoes_necessarias = {"ler", "editar", "excluir"}
```

Cada usuário possui um conjunto de permissões:

```python
usuarios = {
    "ana": {"ler", "editar", "excluir"},
    "bruno": {"ler", "editar"},
    "carlos": {"ler", "excluir"},
    "diana": {"ler", "editar", "excluir", "comentar"}
}
```

Crie um dicionário chamado `acessos`, em que a chave é o nome do usuário e o valor é `"permitido"` ou `"negado"`.


````python
# Gabarito

permissoes_necessarias = {"ler", "editar", "excluir"}

usuarios = {
    "ana": {"ler", "editar", "excluir"},
    "bruno": {"ler", "editar"},
    "carlos": {"ler", "excluir"},
    "diana": {"ler", "editar", "excluir", "comentar"}
}

acessos = {}

for usuario, permissoes in usuarios.items():
    if permissoes_necessarias.issubset(permissoes):
        acessos[usuario] = "permitido"
    else:
        acessos[usuario] = "negado"

acessos
````

## Exercício 8 — Descobrindo permissões faltantes

Usando os mesmos dados do exercício anterior, crie um dicionário chamado `permissoes_faltantes`.

Nesse dicionário:

- cada chave deve ser o nome do usuário;
- cada valor deve ser o conjunto de permissões que faltam para esse usuário acessar a área administrativa.

Usuários que já possuem todas as permissões devem aparecer com conjunto vazio.


````python
# Gabarito

permissoes_necessarias = {"ler", "editar", "excluir"}

usuarios = {
    "ana": {"ler", "editar", "excluir"},
    "bruno": {"ler", "editar"},
    "carlos": {"ler", "excluir"},
    "diana": {"ler", "editar", "excluir", "comentar"}
}

permissoes_faltantes = {}

for usuario, permissoes in usuarios.items():
    faltantes = permissoes_necessarias.difference(permissoes)
    permissoes_faltantes[usuario] = faltantes

permissoes_faltantes
````

## Exercício 9 — Limpando e-mails duplicados

Você recebeu uma lista de e-mails com repetições, espaços extras e diferenças entre maiúsculas e minúsculas.

```python
emails = [
    "Ana@Email.com",
    " bruno@email.com ",
    "ana@email.com",
    "CARLOS@email.com ",
    "bruno@email.com",
    "diana@email.com"
]
```

Crie uma lista com os e-mails únicos, todos em minúsculas e sem espaços extras.

Depois, crie um dicionário chamado `relatorio_emails` contendo:

- `"emails_unicos"`: lista de e-mails únicos;
- `"quantidade_original"`: quantidade original de e-mails;
- `"quantidade_final"`: quantidade depois da limpeza.


````python
# Gabarito

emails = [
    "Ana@Email.com",
    " bruno@email.com ",
    "ana@email.com",
    "CARLOS@email.com ",
    "bruno@email.com",
    "diana@email.com"
]

emails_limpos = []

for email in emails:
    email_limpo = email.strip().lower()
    emails_limpos.append(email_limpo)

emails_unicos = list(set(emails_limpos))

relatorio_emails = {
    "emails_unicos": emails_unicos,
    "quantidade_original": len(emails),
    "quantidade_final": len(emails_unicos)
}

relatorio_emails
````

## Exercício 10 — Tags únicas por postagem

Uma rede social armazenou postagens com suas respectivas tags.

```python
postagens = [
    {"usuario": "ana", "tags": ["python", "dados", "python"]},
    {"usuario": "bruno", "tags": ["sql", "dados"]},
    {"usuario": "carlos", "tags": ["python", "web", "sql"]},
    {"usuario": "diana", "tags": ["excel", "dados"]}
]
```

Crie um conjunto chamado `todas_as_tags` com todas as tags diferentes usadas nas postagens.

Depois, crie um dicionário chamado `resumo_tags` contendo:

- `"tags"`: o conjunto de tags únicas;
- `"quantidade_tags"`: quantidade de tags diferentes.


````python
# Gabarito

postagens = [
    {"usuario": "ana", "tags": ["python", "dados", "python"]},
    {"usuario": "bruno", "tags": ["sql", "dados"]},
    {"usuario": "carlos", "tags": ["python", "web", "sql"]},
    {"usuario": "diana", "tags": ["excel", "dados"]}
]

todas_as_tags = set()

for postagem in postagens:
    for tag in postagem["tags"]:
        todas_as_tags.add(tag)

resumo_tags = {
    "tags": todas_as_tags,
    "quantidade_tags": len(todas_as_tags)
}

resumo_tags
````

## Exercício 11 — Filtrando produtos por categorias permitidas

Uma loja possui uma lista de produtos. Algumas categorias não são permitidas no relatório final.

```python
produtos = [
    {"nome": "Camisa Azul", "categoria": "camisa"},
    {"nome": "Tênis Casual", "categoria": "calçado"},
    {"nome": "Boné", "categoria": "acessório"},
    {"nome": "Calça Jeans", "categoria": "calça"},
    {"nome": "Relógio", "categoria": "acessório"}
]

categorias_permitidas = {"camisa", "calça", "calçado"}
```

Crie uma nova lista chamada `produtos_filtrados`, contendo apenas os produtos cuja categoria está em `categorias_permitidas`.


````python
# Gabarito

produtos = [
    {"nome": "Camisa Azul", "categoria": "camisa"},
    {"nome": "Tênis Casual", "categoria": "calçado"},
    {"nome": "Boné", "categoria": "acessório"},
    {"nome": "Calça Jeans", "categoria": "calça"},
    {"nome": "Relógio", "categoria": "acessório"}
]

categorias_permitidas = {"camisa", "calça", "calçado"}

produtos_filtrados = []

for produto in produtos:
    if produto["categoria"] in categorias_permitidas:
        produtos_filtrados.append(produto)

produtos_filtrados
````

## Exercício 12 — Resumo de respostas de pesquisa

Uma pesquisa perguntou quais ferramentas cada aluno já usou.

```python
respostas = {
    "Ana": ["Python", "SQL", "Excel", "Python"],
    "Bruno": ["Excel", "Power BI", "SQL"],
    "Carlos": ["Python", "Tableau"],
    "Diana": ["SQL", "Excel", "Power BI", "SQL"]
}
```

Crie um dicionário chamado `resumo_pesquisa` contendo:

- `"ferramentas_por_aluno"`: um dicionário em que cada aluno aponta para um conjunto de ferramentas únicas;
- `"todas_ferramentas"`: conjunto com todas as ferramentas citadas na pesquisa;
- `"quantidade_total_ferramentas"`: quantidade de ferramentas diferentes citadas.


````python
# Gabarito

respostas = {
    "Ana": ["Python", "SQL", "Excel", "Python"],
    "Bruno": ["Excel", "Power BI", "SQL"],
    "Carlos": ["Python", "Tableau"],
    "Diana": ["SQL", "Excel", "Power BI", "SQL"]
}

ferramentas_por_aluno = {}
todas_ferramentas = set()

for aluno, ferramentas in respostas.items():
    ferramentas_unicas = set(ferramentas)
    ferramentas_por_aluno[aluno] = ferramentas_unicas
    todas_ferramentas.update(ferramentas_unicas)

resumo_pesquisa = {
    "ferramentas_por_aluno": ferramentas_por_aluno,
    "todas_ferramentas": todas_ferramentas,
    "quantidade_total_ferramentas": len(todas_ferramentas)
}

resumo_pesquisa
````

## Observação final

Nesta lista, `set` foi usado principalmente como ferramenta prática para:

- remover duplicatas;
- comparar listas;
- limpar dados;
- validar permissões;
- gerar relatórios;
- extrair categorias ou valores únicos.

Na prática, é muito comum usar `set` junto com listas e dicionários, especialmente quando queremos transformar dados repetidos em informações mais organizadas.
