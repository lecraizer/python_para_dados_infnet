# Aula — Abertura de Arquivos em Python

Este notebook é focado apenas em **abrir, ler e escrever arquivos em Python**.

A ideia é entender:

- o que é abrir um arquivo;
- caminhos de arquivos;
- modos de abertura;
- `open()` manual;
- `with open(...)`;
- leitura com `read()`, `readline()` e `readlines()`;
- escrita com `write()`;
- adição de conteúdo com modo append;
- alguns exemplos básicos no final.

Aqui o foco não é JSON, CSV ou Pandas. O foco é a lógica geral de abrir arquivos.


## 1. O que significa abrir um arquivo?

Quando usamos Python, os dados nem sempre estão escritos diretamente dentro do código.

Muitas vezes eles estão em arquivos externos, como:

- `.txt`;
- `.csv`;
- `.json`;
- `.html`;
- `.log`;
- entre outros.

Abrir um arquivo significa pedir ao Python para acessar esse arquivo.

Depois de aberto, podemos fazer operações como:

- ler o conteúdo;
- escrever um novo conteúdo;
- acrescentar informações ao final;
- percorrer linha por linha;
- transformar o conteúdo em listas, strings ou dicionários.


## 2. A função `open()`

A função principal para abrir arquivos em Python é `open()`.

A estrutura básica é:

```python
arquivo = open("nome_do_arquivo.txt", "modo")
```

Exemplo:

```python
arquivo = open("dados.txt", "r")
```

Nesse caso:

- `"dados.txt"` é o nome do arquivo;
- `"r"` indica o modo de abertura;
- `"r"` significa leitura, do inglês *read*.


## 3. Modos de abertura de arquivo

O segundo argumento da função `open()` indica o que queremos fazer com o arquivo.

| Modo | Significado |
|---|---|
| `"r"` | leitura |
| `"w"` | escrita, apagando o conteúdo anterior |
| `"a"` | escrita ao final do arquivo |
| `"x"` | criação de arquivo novo |
| `"r+"` | leitura e escrita |
| `"b"` | modo binário, usado junto com outros modos |

Os modos mais importantes no começo são:

```python
"r"  # ler
"w"  # escrever, substituindo o conteúdo anterior
"a"  # acrescentar ao final
```


## 4. Criando um arquivo de exemplo

Antes de ler um arquivo, precisamos ter um arquivo para testar.

Vamos criar um arquivo simples chamado `mensagem.txt`.


````python
arquivo = open("mensagem.txt", "w", encoding="utf-8")

arquivo.write("Olá, mundo!\n")
arquivo.write("Esta é uma aula sobre arquivos em Python.\n")
arquivo.write("Estamos aprendendo a abrir, ler e escrever arquivos.\n")

arquivo.close()
````

Observe que usamos:

```python
arquivo.close()
```

Quando abrimos um arquivo manualmente, precisamos fechá-lo ao final.

Isso libera o recurso usado pelo sistema operacional.


## 5. Abrindo arquivo de forma manual

Uma forma possível de abrir arquivos é usando `open()` diretamente e depois fechando manualmente com `close()`.

Essa forma tem a seguinte estrutura:

```python
arquivo = open("mensagem.txt", "r", encoding="utf-8")
conteudo = arquivo.read()
arquivo.close()
```

Ela não usa `if`.

O que acontece aqui é que o arquivo aberto fica guardado na variável `arquivo`.

Depois usamos `.read()` para ler seu conteúdo e `.close()` para fechar.


````python
arquivo = open("mensagem.txt", "r", encoding="utf-8")

conteudo = arquivo.read()

arquivo.close()

print(conteudo)
````

Essa forma funciona.

Mas ela tem um problema: se esquecermos de usar `close()`, o arquivo pode ficar aberto desnecessariamente.

Por isso, em geral, usamos uma forma mais segura: `with open(...)`.


## 5.1. Forma manual completa: abrir, ler, usar e fechar

A forma manual segue esta lógica:

```python
arquivo = open("arquivo.txt", "r", encoding="utf-8")
conteudo = arquivo.read()
arquivo.close()
```

Ou seja:

1. `open()` abre o arquivo;
2. o arquivo fica guardado em uma variável;
3. usamos métodos como `.read()` para ler o conteúdo;
4. depois fechamos o arquivo com `.close()`.

Essa forma é útil para entender o processo completo.


````python
arquivo = open("mensagem.txt", "r", encoding="utf-8")

conteudo = arquivo.read()

arquivo.close()

print(conteudo)
````

Depois do `close()`, o arquivo foi fechado.

A variável `conteudo` continua existindo, porque ela guardou o texto que foi lido.

Mas o arquivo em si não deve mais ser usado depois de fechado.


````python
print(conteudo)
````

### O que acontece se tentarmos ler depois de fechar?

Depois de fechar o arquivo, tentar usar `.read()` novamente gera erro.

O exemplo abaixo está comentado para não interromper a execução do notebook:

```python
arquivo.read()
```

Isso acontece porque o arquivo já foi fechado.


## 5.2. Então quando usar a forma manual?

A forma manual funciona, mas exige cuidado.

Ela pode ser útil para fins didáticos, porque mostra claramente as etapas:

```python
arquivo = open(...)
conteudo = arquivo.read()
arquivo.close()
```

Porém, em código real, normalmente preferimos `with open(...)`, porque ele fecha o arquivo automaticamente.


## 5.3. Comparação direta

### Forma manual

```python
arquivo = open("mensagem.txt", "r", encoding="utf-8")
conteudo = arquivo.read()
arquivo.close()
```

### Forma recomendada

```python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()
```

Nas duas formas, o conteúdo pode ser usado depois:

```python
print(conteudo)
```

A diferença é que, com `with`, o Python fecha o arquivo automaticamente ao sair do bloco.


## 6. A forma recomendada: `with open(...)`

A forma mais comum e recomendada é:

```python
with open("arquivo.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()
```

O `with` cria um bloco de código.

Quando o bloco termina, o Python fecha o arquivo automaticamente.

Isso evita a necessidade de chamar `arquivo.close()` manualmente.


````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

print(conteudo)
````

## 7. Comparando as duas formas

### Forma manual

```python
arquivo = open("mensagem.txt", "r")
conteudo = arquivo.read()
arquivo.close()
```

### Forma com `with`

```python
with open("mensagem.txt", "r") as arquivo:
    conteudo = arquivo.read()
```

A segunda forma é mais segura e mais usada em códigos reais.

Mesmo assim, é importante conhecer a primeira, porque ela ajuda a entender o que está acontecendo por trás.


## 8. O parâmetro `encoding`

Em muitos casos, usamos:

```python
encoding="utf-8"
```

Exemplo:

```python
with open("arquivo.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()
```

O `encoding` define como os caracteres do arquivo devem ser interpretados.

Isso é especialmente importante para textos com acentos, como:

- ação;
- coração;
- município;
- introdução;
- João.

Para a maior parte dos arquivos de texto atuais, `utf-8` é uma boa escolha.


## 9. Lendo o arquivo inteiro com `read()`

O método `read()` lê todo o conteúdo do arquivo de uma vez.


````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

conteudo
````

Quando usamos `print()`, as quebras de linha aparecem de forma mais natural.


````python
print(conteudo)
````

## 10. Lendo uma linha com `readline()`

O método `readline()` lê uma linha por vez.


````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    primeira_linha = arquivo.readline()
    segunda_linha = arquivo.readline()

print(primeira_linha)
print(segunda_linha)
````

Cada chamada de `readline()` avança para a próxima linha do arquivo.


## 11. Lendo todas as linhas com `readlines()`

O método `readlines()` lê todas as linhas e retorna uma lista.

Cada elemento da lista é uma linha do arquivo.


````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    linhas = arquivo.readlines()

linhas
````

Observe que as linhas podem vir com o caractere `\n`, que representa quebra de linha.


## 12. Percorrendo um arquivo linha por linha

Uma forma muito comum de trabalhar com arquivos é usar um `for`.

Exemplo:


````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        print(linha)
````

Às vezes aparecem linhas em branco extras por causa do `\n` que já existe no arquivo e do `print()` que também pula linha.

Podemos usar `.strip()` para remover espaços e quebras de linha no começo e no fim.


````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        print(linha.strip())
````

## 13. Escrevendo em arquivos com modo `"w"`

O modo `"w"` serve para escrever em um arquivo.

Atenção: se o arquivo já existir, o conteúdo anterior será apagado.


````python
with open("novo_arquivo.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write("Primeira linha do arquivo.\n")
    arquivo.write("Segunda linha do arquivo.\n")
````

Agora vamos abrir o arquivo criado para conferir o conteúdo.


````python
with open("novo_arquivo.txt", "r", encoding="utf-8") as arquivo:
    print(arquivo.read())
````

## 14. Acrescentando conteúdo com modo `"a"`

O modo `"a"` vem de *append*.

Ele serve para acrescentar conteúdo ao final do arquivo, sem apagar o que já existia.


````python
with open("novo_arquivo.txt", "a", encoding="utf-8") as arquivo:
    arquivo.write("Terceira linha acrescentada depois.\n")
````

````python
with open("novo_arquivo.txt", "r", encoding="utf-8") as arquivo:
    print(arquivo.read())
````

## 15. Criando arquivo apenas se ele não existir com `"x"`

O modo `"x"` cria um arquivo novo.

Se o arquivo já existir, o Python gera erro.

É útil quando queremos evitar sobrescrever um arquivo por acidente.

Exemplo:

```python
with open("arquivo_novo.txt", "x", encoding="utf-8") as arquivo:
    arquivo.write("Conteúdo inicial")
```

Se `arquivo_novo.txt` já existir, esse código não sobrescreve o arquivo.


## 16. Caminhos de arquivos

Até agora usamos arquivos no mesmo local do notebook.

Exemplo:

```python
with open("mensagem.txt", "r") as arquivo:
    ...
```

Mas também podemos usar caminhos.

Exemplo em uma pasta:

```python
with open("dados/mensagem.txt", "r") as arquivo:
    ...
```

Em sistemas Linux ou Mac, caminhos usam `/`.

Em Windows, muitas vezes aparecem caminhos com `\`, mas o Python também aceita `/` em muitos casos.

Exemplo:

```python
with open("C:/Users/Luis/Downloads/arquivo.txt", "r") as arquivo:
    ...
```


## 17. Arquivos como texto

Quando abrimos um arquivo `.txt`, o conteúdo lido normalmente vem como string.

Exemplo:

```python
conteudo = arquivo.read()
```

Nesse caso, `conteudo` é uma string.

Isso significa que podemos usar métodos de string, como:

- `.lower()`;
- `.upper()`;
- `.strip()`;
- `.split()`;
- `.replace()`;
- `.startswith()`;
- `.endswith()`.


````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

print(type(conteudo))
print(conteudo.upper())
````

## 18. Transformando linhas em listas

Se cada linha do arquivo representa um item, podemos transformar o arquivo em uma lista.

Vamos criar um arquivo simples com nomes.


````python
with open("nomes.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write("Ana\n")
    arquivo.write("Bruno\n")
    arquivo.write("Carlos\n")
    arquivo.write("Diana\n")
````

````python
nomes = []

with open("nomes.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        nomes.append(linha.strip())

nomes
````

## 19. Contando informações em um arquivo

Arquivos podem ser usados junto com listas e dicionários.

Vamos criar um arquivo com tipos sanguíneos.


````python
with open("tipos_sanguineos.txt", "w", encoding="utf-8") as arquivo:
    arquivo.write("A\n")
    arquivo.write("O\n")
    arquivo.write("B\n")
    arquivo.write("A\n")
    arquivo.write("AB\n")
    arquivo.write("O\n")
    arquivo.write("A\n")
````

Agora vamos abrir o arquivo e contar quantas vezes cada tipo aparece.


````python
contagem = {}

with open("tipos_sanguineos.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        tipo = linha.strip()

        if tipo not in contagem:
            contagem[tipo] = 0

        contagem[tipo] += 1

contagem
````

Esse tipo de lógica é muito comum:

1. abrir arquivo;
2. percorrer linha por linha;
3. limpar a linha com `.strip()`;
4. extrair a informação importante;
5. guardar o resultado em lista, set ou dicionário.


## 20. Boas práticas

Algumas boas práticas ao trabalhar com arquivos:

- preferir `with open(...)`;
- informar `encoding="utf-8"` quando trabalhar com texto;
- usar nomes de arquivos claros;
- fechar arquivos quando usar abertura manual;
- verificar se o arquivo está na pasta correta;
- usar `.strip()` para limpar quebras de linha;
- tomar cuidado com o modo `"w"`, pois ele apaga o conteúdo anterior.


# Exemplos básicos finais

A seguir, alguns exemplos curtos para revisar.


## Exemplo 1 — Ler arquivo inteiro

````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    conteudo = arquivo.read()

print(conteudo)
````

## Exemplo 2 — Ler linha por linha

````python
with open("mensagem.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        print(linha.strip())
````

## Exemplo 3 — Criar uma lista a partir de um arquivo

````python
nomes = []

with open("nomes.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        nomes.append(linha.strip())

nomes
````

## Exemplo 4 — Escrever uma lista em um arquivo

````python
frutas = ["maçã", "banana", "uva", "laranja"]

with open("frutas.txt", "w", encoding="utf-8") as arquivo:
    for fruta in frutas:
        arquivo.write(fruta + "\n")
````

## Exemplo 5 — Contar linhas de um arquivo

````python
quantidade = 0

with open("frutas.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        quantidade += 1

print("Quantidade de linhas:", quantidade)
````

## Exemplo 6 — Contar ocorrências em um arquivo

````python
contagem = {}

with open("tipos_sanguineos.txt", "r", encoding="utf-8") as arquivo:
    for linha in arquivo:
        valor = linha.strip()
        contagem[valor] = contagem.get(valor, 0) + 1

contagem
````
