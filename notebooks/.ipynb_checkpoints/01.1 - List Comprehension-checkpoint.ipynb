## 1. O que são list comprehensions?

**List comprehensions** são uma forma mais curta, elegante e idiomática de criar listas em Python.

Em vez de escrever um `for` tradicional e usar `append()` a cada iteração, podemos construir a lista diretamente em uma única expressão.

Essa estrutura é muito comum em Python e aparece com frequência em códigos reais. Por isso, vale a pena aprender não só **como usar**, mas também **quando usar**.


## 2. Motivação

Muitas vezes queremos:

- percorrer uma coleção
- transformar cada elemento
- ou filtrar alguns elementos
- e gerar uma nova lista no final

Isso aparece o tempo todo em programação.

### Exemplo de necessidade comum:
> “Quero pegar uma lista de números e criar outra lista com os quadrados desses números.”

Vamos começar pela forma tradicional.


````python
numeros = [1, 2, 3, 4, 5]
quadrados = []

for n in numeros:
    quadrados.append(n ** 2)

print(quadrados)
````

Agora a mesma ideia com **list comprehension**:

````python
numeros = [1, 2, 3, 4, 5]
quadrados = [n ** 2 for n in numeros]

print(quadrados)
````

## 3. Estrutura geral

A forma básica é:

```python
[expressao for item in iteravel]
```

Lendo em português:

> “Crie uma lista com `expressao` para cada `item` em `iteravel`.”

### Exemplo:
```python
[n * 2 for n in [1, 2, 3]]
```

Resultado:
```python
[2, 4, 6]
```


## 4. Por que usar?

### Código mais conciso
Com menos linhas, conseguimos expressar a mesma ideia.

### Leitura mais direta
Quando bem usada, a list comprehension deixa claro que queremos **gerar uma nova lista** a partir de outra estrutura.

### Estilo mais idiomático
É uma construção muito comum em Python. Aprender isso ajuda a ler e escrever código mais próximo do que programadores Python costumam fazer.

### Pode ser mais eficiente
Em muitos casos, list comprehensions têm desempenho um pouco melhor do que `for` com `append()`.


## 5. Exemplo simples de transformação

Vamos multiplicar todos os valores por 2.


````python
nums = [1, 2, 3, 4]
dobro = [n * 2 for n in nums]

print(dobro)
````

## 6. Exemplo com filtro (`if`)

Também podemos usar uma condição para selecionar apenas alguns elementos.

A estrutura fica assim:

```python
[expressao for item in iteravel if condicao]
```

Exemplo: pegar apenas os números pares.


````python
pares = [n for n in range(10) if n % 2 == 0]
print(pares)
````

Agora um exemplo com **transformação + filtro**:


````python
quadrados_pares = [n ** 2 for n in range(10) if n % 2 == 0]
print(quadrados_pares)
````

## 7. Exemplos com strings

List comprehensions também funcionam muito bem com strings e listas de strings.


````python
palavra = "python"
letras_maiusculas = [letra.upper() for letra in palavra]

print(letras_maiusculas)
````

````python
nomes = ["ana", "joao", "maria"]
nomes_formatados = [nome.capitalize() for nome in nomes]

print(nomes_formatados)
````

````python
strings = ["  oi ", " tchau  ", "  beleza"]
limpas = [s.strip() for s in strings]

print(limpas)
````

## 8. Condicional inline: `if ... else`

Há um caso importante em que queremos escolher entre dois valores diferentes para cada elemento.

A estrutura fica assim:

```python
[valor_if if condicao else valor_else for item in iteravel]
```

Exemplo: classificar números como `"par"` ou `"impar"`.


````python
resultado = ["par" if n % 2 == 0 else "impar" for n in range(6)]
print(resultado)
````

Outro exemplo:
- se o número for par, mantém o número
- se for ímpar, multiplica por 10


````python
nums = [1, 2, 3, 4, 5]
resultado = [n if n % 2 == 0 else n * 10 for n in nums]

print(resultado)
````

## 9. Comparação mental com `for`

É útil enxergar a equivalência entre as duas formas.

### `for` tradicional
```python
nova_lista = []
for x in lista:
    if condicao:
        nova_lista.append(f(x))
```

### list comprehension
```python
[f(x) for x in lista if condicao]
```

Essa comparação ajuda muito os alunos a entenderem de onde vem a sintaxe.


## 10. Padrões comuns

### 1) Apenas transformar
```python
[x * 2 for x in lista]
```

### 2) Apenas filtrar
```python
[x for x in lista if condicao]
```

### 3) Transformar e filtrar
```python
[x * 2 for x in lista if condicao]
```

Esses três padrões já cobrem boa parte do uso inicial da ferramenta.


## 11. Exemplo com listas dentro de listas

Às vezes temos uma matriz, ou seja, uma lista que contém outras listas.


````python
matriz = [
    [1, 2, 3],
    [4, 5, 6]
]

flatten = [num for linha in matriz for num in linha]
print(flatten)
````

Nesse caso, a ordem importa:

```python
[num for linha in matriz for num in linha]
```

Isso equivale a:

```python
for linha in matriz:
    for num in linha:
        ...
```

Esse tipo de construção é útil, mas costuma ser melhor deixar para depois que os alunos estiverem confortáveis com os casos simples.


## 12. Quando usar e quando evitar

### Vale a pena usar quando:
- a lógica é simples
- a intenção é gerar uma nova lista
- o código fica mais limpo que um `for` tradicional

### Melhor evitar quando:
- há muitas condições
- a expressão fica grande demais
- a leitura fica confusa
- a lógica é complexa

Regra prática:

> Se a list comprehension ficou difícil de ler, provavelmente é melhor usar um `for`.


## 13. Erros comuns

### Erro 1: ordem errada
```python
[x for if x > 2 in lista]
```

### Erro 2: usar `if` no lugar errado
```python
[x if x > 2 for x in lista]
```

### Erro 3: confundir filtro com `if/else`
```python
[x for x in lista if condicao else outro_valor]
```

O formato correto depende do objetivo:

- **filtrar**:
```python
[x for x in lista if condicao]
```

- **escolher entre dois valores**:
```python
[x if condicao else y for x in lista]
```


## 14. Por que isso é importante?

Aprender list comprehensions é importante porque:

- melhora a leitura de códigos Python
- reduz repetição
- introduz uma forma mais expressiva de transformar dados
- prepara terreno para **dict comprehensions**, **set comprehensions** e até ideias de programação funcional

Ou seja, não é só uma “forma curta de escrever”. É também uma maneira de pensar melhor sobre transformação de dados.


## 15. Gancho para a próxima aula: dicionários

Assim como existe list comprehension, também existe **dict comprehension**.

Exemplo:


````python
quadrados = {n: n ** 2 for n in range(5)}
print(quadrados)
````

Isso abre naturalmente a próxima parte da disciplina: **dicionários em Python**.

## 16. Resumo final

List comprehensions servem para criar listas de forma mais curta e elegante.

### Estruturas principais:
```python
[expressao for item in iteravel]
[expressao for item in iteravel if condicao]
[valor_if if condicao else valor_else for item in iteravel]
```

### Ideias centrais:
- transformar
- filtrar
- escrever código mais idiomático
- manter a legibilidade

### Regra de ouro:
> Use quando deixar o código mais claro, não apenas mais curto.


## 17. Bloco extra para exploração em aula

````python
numeros = [1, 2, 3, 4, 5, 6]

print([n * 3 for n in numeros])
print([n for n in numeros if n > 3])
print([n ** 2 for n in numeros if n % 2 == 1])
print(["par" if n % 2 == 0 else "impar" for n in numeros])
````
