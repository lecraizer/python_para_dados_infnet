# 🧠 Dicionários em Python

Este notebook apresenta a parte teórica sobre **dicionários (`dict`)** em Python, com foco em:

- o que são dicionários
- como acessar, alterar e percorrer dados
- possíveis `for` com dicionários
- aplicações importantes:
  - contagem de frequência
  - agrupamento de dados
  - transformação de dados

---


## 1. O que é um dicionário?

Um **dicionário** é uma estrutura que armazena pares no formato:

**chave → valor**

Diferente de uma lista, em que acessamos os dados por posição, no dicionário acessamos por **chave**.

### Exemplo:


````python
pessoa = {
    "nome": "Luis",
    "idade": 30,
    "cidade": "Rio de Janeiro"
}

pessoa
````

Nesse caso:

- `"nome"`, `"idade"` e `"cidade"` são as **chaves**
- `"Luis"`, `30` e `"Rio de Janeiro"` são os **valores**

---


## 2. Comparação com listas

### Lista
Uma lista guarda valores em sequência e é acessada por índice:


````python
lista = ["ana", "joao", "maria"]

print(lista[0])
print(lista[1])
````

### Dicionário
Um dicionário guarda valores associados a chaves:


````python
idades = {
    "ana": 20,
    "joao": 25,
    "maria": 22
}

print(idades["ana"])
print(idades["maria"])
````

Ou seja:

- **lista** → acesso por posição
- **dicionário** → acesso por chave

---


## 3. Características importantes

### 3.1. As chaves são únicas
Se uma chave aparecer repetida, a última sobrescreve a anterior.


````python
d = {
    "a": 10,
    "a": 50
}

d
````

### 3.2. As chaves devem ser imutáveis

Normalmente usamos:

- strings
- números
- tuplas

Não usamos listas ou dicionários como chave.

### 3.3. Os valores podem ser de qualquer tipo


````python
exemplo = {
    "nome": "Ana",
    "idade": 21,
    "notas": [8.0, 7.5, 9.0],
    "aprovada": True
}

exemplo
````

---
## 4. Acessando valores


````python
pessoa = {
    "nome": "Luis",
    "idade": 30
}

print(pessoa["nome"])
print(pessoa["idade"])
````

Se a chave não existir, Python gera erro.

Uma forma mais segura é usar `.get()`:


````python
print(pessoa.get("nome"))
print(pessoa.get("altura"))      # retorna None
print(pessoa.get("altura", 0))   # valor padrão
````

---
## 5. Alterando e adicionando valores


````python
pessoa = {
    "nome": "Luis",
    "idade": 30
}

# alterar valor existente
pessoa["idade"] = 31

# adicionar novo par chave-valor
pessoa["cidade"] = "Rio"

pessoa
````

---
## 6. Removendo elementos


````python
pessoa = {
    "nome": "Luis",
    "idade": 30,
    "cidade": "Rio"
}

del pessoa["cidade"]
pessoa
````

````python
pessoa = {
    "nome": "Luis",
    "idade": 30,
    "cidade": "Rio"
}

idade_removida = pessoa.pop("idade")

print("Idade removida:", idade_removida)
print(pessoa)
````

---
## 7. Possíveis `for` com dicionários

Essa é uma das partes mais importantes.


### 7.1. Percorrendo as chaves
Quando usamos um `for` diretamente no dicionário, ele percorre as **chaves**.


````python
notas = {
    "Ana": 8.5,
    "Bruno": 7.0,
    "Carla": 9.2
}

for chave in notas: # or .keys()
    print(chave)
````

### 7.2. Percorrendo os valores
Para percorrer só os valores, usamos `.values()`.


````python
for valor in notas.values():
    print(valor)
````

### 7.3. Percorrendo chave e valor
Esse é o formato mais usado. Para isso, usamos `.items()`.


````python
for chave, valor in notas.items():
    print(chave, "->", valor)
````

### 7.4. Usando condição dentro do `for`
Também podemos combinar com `if`.


````python
for aluno, nota in notas.items():
    if nota >= 7:
        print(aluno, "foi aprovado")
````

## 8. Aplicação 1 — Contagem de frequência

Uma das aplicações mais clássicas dos dicionários é **contar quantas vezes algo aparece**.

### Exemplo 1: contar letras em uma palavra


````python
palavra = "banana"
contagem = {}

for letra in palavra:
    if letra in contagem:
        contagem[letra] += 1
    else:
        contagem[letra] = 1

contagem
````

### Como esse algoritmo funciona?

Para cada letra:

- se ela já existe no dicionário, aumentamos a contagem
- se ainda não existe, começamos com 1

### Exemplo 2: contar números em uma lista


````python
numeros = [1, 2, 1, 3, 2, 1, 4, 3, 2]
frequencia = {}

for n in numeros:
    if n in frequencia:
        frequencia[n] += 1
    else:
        frequencia[n] = 1

frequencia
````

### Exemplo 3: contar palavras em uma frase


````python
frase = "python é legal e python é útil"
palavras = frase.split()

contagem_palavras = {}

for palavra in palavras:
    if palavra in contagem_palavras:
        contagem_palavras[palavra] += 1
    else:
        contagem_palavras[palavra] = 1

contagem_palavras
````

## 9. Aplicação 2 — Agrupar dados

Outra aplicação muito importante é **agrupar elementos por alguma característica**.

### Exemplo 1: agrupar nomes pela inicial


````python
nomes = ["ana", "amanda", "bruno", "bia", "carlos", "caio"]
grupos = {}

for nome in nomes:
    inicial = nome[0]

    if inicial not in grupos:
        grupos[inicial] = []

    grupos[inicial].append(nome)

grupos
````

Nesse exemplo:

- cada chave é uma inicial
- cada valor é uma lista de nomes que começam com essa letra

### Exemplo 2: agrupar números em pares e ímpares


````python
numeros = [1, 2, 3, 4, 5, 6, 7, 8]
grupos = {
    "pares": [],
    "impares": []
}

for n in numeros:
    if n % 2 == 0:
        grupos["pares"].append(n)
    else:
        grupos["impares"].append(n)

grupos
````

### Exemplo 3: agrupar alunos por situação


````python
notas = {
    "Ana": 8.5,
    "Bruno": 5.0,
    "Carla": 7.2,
    "Diego": 6.4
}

situacao = {
    "aprovados": [],
    "reprovados": []
}

for aluno, nota in notas.items():
    if nota >= 7:
        situacao["aprovados"].append(aluno)
    else:
        situacao["reprovados"].append(aluno)

situacao
````

## 10. Aplicação 3 — Transformação de dados

Dicionários também são ótimos para **transformar dados**, reorganizando informações.

### Exemplo 1: transformar lista de nomes em dicionário com tamanho


````python
nomes = ["Ana", "Bruno", "Carla"]
resultado = {}

for nome in nomes:
    resultado[nome] = len(nome)

resultado
````

Agora cada nome virou uma chave, e o valor associado é seu tamanho.

### Exemplo 2: transformar lista de números em dicionário com quadrados


````python
numeros = [1, 2, 3, 4, 5]
quadrados = {}

for n in numeros:
    quadrados[n] = n ** 2

quadrados
````

### Exemplo 3: transformar notas numéricas em situação textual


````python
notas = {
    "Ana": 8.0,
    "Bruno": 6.5,
    "Carla": 9.1
}

situacoes = {}

for aluno, nota in notas.items():
    if nota >= 7:
        situacoes[aluno] = "aprovado"
    else:
        situacoes[aluno] = "reprovado"

situacoes
````

### Exemplo 4: criar um dicionário a partir de duas listas


````python
nomes = ["Ana", "Bruno", "Carla"]
idades = [20, 25, 22]

cadastro = {}

for i in range(len(nomes)):
    cadastro[nomes[i]] = idades[i]

cadastro
````

## 11. Por que usar dicionários?

Dicionários são muito úteis quando queremos:

- associar informações a nomes, códigos ou categorias
- fazer buscas rápidas
- contar ocorrências
- agrupar elementos
- transformar dados em uma estrutura mais organizada

Em muitos problemas reais, uma lista sozinha não é suficiente.  
O dicionário aparece justamente quando precisamos de uma estrutura **mais inteligente para organizar os dados**.

---


## 12. Erros comuns dos alunos

### 1. Confundir chave com índice
Lista usa posição. Dicionário usa chave.

### 2. Tentar acessar uma chave que não existe
Isso gera erro se usar colchetes diretamente.

### 3. Esquecer `.items()` ao querer chave e valor
Errado:
```python
for k, v in d:
    ...
```

Certo:
```python
for k, v in d.items():
    ...
```

### 4. Sobrescrever sem perceber
Se repetir uma chave, o valor anterior é perdido.

---


## 13. Resumo final

Dicionários são estruturas de dados baseadas em:

**chave → valor**

Eles são especialmente bons para:

- representar cadastros
- contar frequências
- agrupar informações
- transformar dados
- acessar rapidamente valores associados a uma chave

Enquanto listas são ótimas para sequências, dicionários são ótimos para **organização e busca por significado**.
