## Exercício 1 - Soma por categoria
Dado a lista de vendas, calcule o total vendido por categoria.

```python
vendas = [
    {"categoria": "roupa", "valor": 100},
    {"categoria": "comida", "valor": 50},
    {"categoria": "roupa", "valor": 70}
]

# Gabarito
resultado = {}
for v in vendas:
    cat = v["categoria"]
    valor = v["valor"]
    if cat not in resultado:
        resultado[cat] = 0
    resultado[cat] += valor

print(resultado)
```

## Exercício 2- Palavra mais frequente
Conte a frequência das palavras e encontre a mais frequente.

```python
palavras = ["a", "b", "a", "c", "b", "a"]

# Gabarito
freq = {}
for p in palavras:
    if p not in freq:
        freq[p] = 0
    freq[p] += 1

mais_freq = None
max_qtd = 0

for palavra, qtd in freq.items():
    if qtd > max_qtd:
        max_qtd = qtd
        mais_freq = palavra

print(freq)
print("Mais frequente:", mais_freq)
```

## Exercício 3 - Produto mais caro por categoria
Encontre o produto mais caro em cada categoria.

```python
produtos = [
    {"nome": "camisa", "categoria": "roupa", "preco": 50},
    {"nome": "calça", "categoria": "roupa", "preco": 120},
    {"nome": "pizza", "categoria": "comida", "preco": 40}
]

# Gabarito
mais_caros = {}

for p in produtos:
    cat = p["categoria"]
    if cat not in mais_caros or p["preco"] > mais_caros[cat]["preco"]:
        mais_caros[cat] = p

print(mais_caros)
```

## Exercício 4 - Normalização de dados
Normalize os nomes e some valores de chaves duplicadas.

```python
dados = {
    "ana": 10,
    "Ana ": 8,
    " BRUNO": 7,
    "bruno": 9
}

# Gabarito
normalizado = {}

for nome, valor in dados.items():
    nome_limpo = nome.strip().capitalize()
    if nome_limpo not in normalizado:
        normalizado[nome_limpo] = 0
    normalizado[nome_limpo] += valor

print(normalizado)
```

## Exercício 5 - Histórico de movimentos
Agrupe os valores por pessoa.

```python
movimentos = [
    ("Ana", 100),
    ("Bruno", 50),
    ("Ana", -30),
    ("Bruno", 20)
]

# Gabarito
historico = {}

for nome, valor in movimentos:
    if nome not in historico:
        historico[nome] = []
    historico[nome].append(valor)

print(historico)
```
