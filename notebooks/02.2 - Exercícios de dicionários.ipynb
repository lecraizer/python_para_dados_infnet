# Exercícios - Dicionários

## 1. Buscar idade

Retorne a idade pelo nome. Se não existir, retorne 'não encontrado'.

```python
def buscar_idade(dados, nome):
    return dados.get(nome, "não encontrado")
```

## 2. Aumentar notas

Aumente todas as notas em 1 ponto.

```python
def aumentar_notas(notas):
    novo = {}
    for aluno, nota in notas.items():
        novo[aluno] = nota + 1
    return novo
```

## 3. Filtrar aprovados

Retorne lista de alunos com nota >= 7.

```python
def filtrar_aprovados(notas):
    aprovados = []
    for aluno, nota in notas.items():
        if nota >= 7:
            aprovados.append(aluno)
    return aprovados
```

## 4. Contar letras

Conte a frequência de cada letra.

```python
def contar_letras(texto):
    contagem = {}
    for letra in texto:
        if letra in contagem:
            contagem[letra] += 1
        else:
            contagem[letra] = 1
    return contagem
```

## 5. Contar números

Conte a frequência dos números.

```python
def contar_numeros(lista):
    freq = {}
    for n in lista:
        if n in freq:
            freq[n] += 1
        else:
            freq[n] = 1
    return freq
```

## 6. Agrupar por inicial

Agrupe nomes pela primeira letra.

```python
def agrupar_por_inicial(nomes):
    grupos = {}
    for nome in nomes:
        inicial = nome[0]
        if inicial not in grupos:
            grupos[inicial] = []
        grupos[inicial].append(nome)
    return grupos
```

## 7. Separar pares e ímpares

Retorne dict com pares e ímpares.

```python
def separar_pares_impares(lista):
    grupos = {"pares": [], "impares": []}
    for n in lista:
        if n % 2 == 0:
            grupos["pares"].append(n)
        else:
            grupos["impares"].append(n)
    return grupos
```

## 8. Situação dos alunos

Transforme notas em aprovado/reprovado.

```python
def situacao_alunos(notas):
    situacoes = {}
    for aluno, nota in notas.items():
        if nota >= 7:
            situacoes[aluno] = "aprovado"
        else:
            situacoes[aluno] = "reprovado"
    return situacoes
```

## 9. Tamanho das palavras

Mapeie palavra -> tamanho.

```python
def tamanho_palavras(lista):
    resultado = {}
    for palavra in lista:
        resultado[palavra] = len(palavra)
    return resultado
```

## 10. Juntar listas

Associe nomes às idades.

```python
def juntar_listas(nomes, idades):
    cadastro = {}
    for i in range(len(nomes)):
        cadastro[nomes[i]] = idades[i]
    return cadastro
```

## 11. Maior nota

Retorne o aluno com maior nota.

```python
def maior_nota(notas):
    maior_aluno = None
    maior_valor = None
    for aluno, nota in notas.items():
        if maior_valor is None or nota > maior_valor:
            maior_valor = nota
            maior_aluno = aluno
    return maior_aluno
```

## 12. Somar valores

Some todos os valores do dicionário.

```python
def somar_valores(dados):
    soma = 0
    for valor in dados.values():
        soma += valor
    return soma
```

## 13. Inverter dicionário

Troque chave por valor (sem repetidos).

```python
def inverter_dict(d):
    invertido = {}
    for chave, valor in d.items():
        invertido[valor] = chave
    return invertido
```

## 14. Contar palavras

Conte frequência das palavras em uma frase.

```python
def contar_palavras(frase):
    palavras = frase.split()
    contagem = {}
    for palavra in palavras:
        if palavra in contagem:
            contagem[palavra] += 1
        else:
            contagem[palavra] = 1
    return contagem
```

## 15. Agrupar por situação

Separe alunos em aprovados e reprovados.

```python
def agrupar_por_situacao(notas):
    grupos = {"aprovados": [], "reprovados": []}
    for aluno, nota in notas.items():
        if nota >= 7:
            grupos["aprovados"].append(aluno)
        else:
            grupos["reprovados"].append(aluno)
    return grupos
```
