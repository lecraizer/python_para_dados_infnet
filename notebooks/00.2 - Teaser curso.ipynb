## 🔤 Manipulação de Strings

```python
texto = "Python para dados"

print(texto[0])     # primeiro caractere
print(texto[-1])    # último caractere
print(texto[0:6])   # slicing
```

```python
frase = "dados sao o novo petroleo"

palavras = frase.split()
print(palavras)

nova = "-".join(palavras)
print(nova)
```

```python
texto = "   python   "
print(texto.strip())

print("eu gosto de java".replace("java", "python"))
```

## 🔁 Listas e List Comprehension

```python
numeros = [1, 2, 3, 4]

quadrados = [n**2 for n in numeros]
print(quadrados)

pares = [n for n in numeros if n % 2 == 0]
print(pares)
```

```python
nomes = ["Ana", "Bruno", "Carlos"]

for i, nome in enumerate(nomes):
    print(i, nome)
```

## 🌐 JSON

```python
import json

dados = '{"nome": "Luis", "idade": 30}'
obj = json.loads(dados)

print(obj["nome"])
```
