## Exercício 1 — Ler JSON e acessar campos

Dado o JSON:
```python
texto = '{"nome": "Ana", "idade": 22, "curso": "Python"}'
```
Converta para dicionário e imprima o nome.

````
# Gabarito
import json
texto = '{"nome": "Ana", "idade": 22, "curso": "Python"}'
dados = json.loads(texto)
print(dados["nome"])
````

## Exercício 2 — Converter dicionário para JSON

Crie um dicionário de produto e converta para string JSON formatada.

````
# Gabarito
import json
produto = {"nome": "Camisa", "preco": 79.9}
print(json.dumps(produto, indent=4, ensure_ascii=False))
````

## Exercício 3 — Contar por categoria (JSON lista)

Dado JSON de produtos, conte quantos por categoria.

````
# Gabarito
import json
texto='[{"cat":"a"},{"cat":"b"},{"cat":"a"}]'
dados=json.loads(texto)
cont={}
for p in dados:
    c=p["cat"]
    cont[c]=cont.get(c,0)+1
cont
````

## Exercício 4 — Filtrar JSON

Retorne apenas itens com valor > 50.

````
# Gabarito
import json
dados=json.loads('[{"v":10},{"v":80},{"v":60}]')
[x for x in dados if x["v"]>50]
````

## Exercício 5 — Lista → JSON → lista

Transforme lista em JSON e volte para lista.

````
# Gabarito
import json
lista=[1,2,3]
texto=json.dumps(lista)
nova=json.loads(texto)
nova
````

## Exercício 6 — Criar JSON em arquivo

Salve um dicionário em arquivo JSON.

````
# Gabarito
import json
dados={"a":1,"b":2}
with open("ex.json","w") as f:
    json.dump(dados,f)
````

## Exercício 7 — Ler JSON arquivo e somar valores

Leia arquivo JSON e some campo `valor`.

````
# Gabarito
import json
with open("ex.json") as f:
    d=json.load(f)
sum(d.values())
````

## Exercício 8 — Extrair valores únicos

Dado JSON com lista, retorne valores únicos.

````
# Gabarito
import json
dados=json.loads('[{"x":1},{"x":2},{"x":1}]')
set([d["x"] for d in dados])
````

## Exercício 9 — Criar dicionário resumo

Crie dicionário com total e média.

````
# Gabarito
nums=[1,2,3]
{"total":sum(nums),"media":sum(nums)/len(nums)}
````

## Exercício 10 — JSON + dicionário combinado

Conte ocorrências e retorne JSON.

````
# Gabarito
import json
lista=["a","b","a"]
cont={}
for x in lista:
    cont[x]=cont.get(x,0)+1
json.dumps(cont)
````
